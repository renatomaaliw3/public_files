## Data Visualization (Seaborn P - 2)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Data

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Distribution Plots

```python
# displot - shows distribution of a univariate set of observation

sns.displot(tips['total_bill'], bins = 30, kde = True, rug = True).set(title = 'Distribution Plot')
```

```python
# jointplot - display bivariate data
# kind = scatter, reg, resid, kde, hex

sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter')
```

```python
# hex

sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
# reg (regression)

sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')
```

```python
# PAIRPLOT

sns.pairplot(tips)
```

```python
# PAIRPLOT 

sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python
# RUGPLOT (not recommended)

sns.rugplot(tips['total_bill'])
```

## Categorical Plot
       - Types: factorplot, boxplot, violinplot, stripplot, swarmplot, barplot, countplot

```python
# Bar Plot

sns.barplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
# Count Plot

sns.countplot(x = 'sex', data = tips)
```

```python
# Boxplot

sns.boxplot(x = 'sex', y = 'total_bill', data = tips, palette = 'rainbow')
```

```python
# BoxPlot

sns.boxplot(data = tips, orient = 'h')
```

```python
# BoxPlot

sns.boxplot(x = 'day', y = 'total_bill', hue = 'smoker', data = tips, palette = 'coolwarm')
```

```python
# VIOLINPLOT

sns.violinplot(x = 'day', y = 'total_bill', data = tips, pallete = 'rainbow')
```

```python
# STRIPPLOT

sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
# SWARMPLOT

sns.swarmplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
# FACTORPLOT (Similar to Bar Plot)

sns.catplot(x = 'sex', y = 'total_bill', data = tips, kind = 'bar')
```

## Matrix Plot

```python
## Matrix Plot - indicates cluster of data

# Data

flights = sns.load_dataset('flights')
flights
```

```python
# Heat Plot

sns.heatmap(tips.corr(), annot = True, cmap = 'rainbow')
```

```python
# Pivot Table

flight_data = flights.pivot_table(values = 'passengers', index = 'month', columns = 'year')
```

```python
# HeatMap

sns.heatmap(flight_data, linecolor = 'white', linewidth = 1)
```

```python
# ClusterMap

sns.clustermap(flight_data)
```

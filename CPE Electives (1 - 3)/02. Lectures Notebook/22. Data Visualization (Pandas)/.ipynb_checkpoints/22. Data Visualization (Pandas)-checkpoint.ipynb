# Data Visualization (Pandas - Line Plot)

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

### Data Generation Example

```python
df = pd.DataFrame(np.random.randn(500), 
     index = pd.date_range('2019-01-01', periods = 500), columns = ['values'])
df
```

### Line plot

- use to display change over time (trends)
- use to display changes in values

```python
df.plot()
```

```python
# Figure Size (Graph Size)

df.plot(figsize = (15, 8))
```

### Loading Iris Dataset from Seaborn

```python
# Loading Iris Dataset in Seaborn

iris = sns.load_dataset('iris')
```

```python
iris
```

```python
# Graph Size

iris.plot(figsize = (10,5))
```

```python
# Making changes to graphs (Title)

iris.plot(figsize = (10,5), title = 'Iris Data')
```

# Data Visualization (Pandas - Bar Plots)

```python
# Drop Non-Numeric Data

data = iris.drop(['species'], axis = 1)
data
```

### Bar Plot (Vertical)

- use to display or compare numbers or measures for different categories

```python
data.iloc[0].plot(kind = 'bar', title = 'First Sample')
```

### Bar Plot (Horizontal)

```python
data.iloc[0].plot(kind = 'barh', title = 'First Sample')
```

# Data Visualization (Stacked Plot)

### Data

```python
df = pd.DataFrame(np.random.randn(10,4), columns = ['a','b','c','d'])
df
```

### Stacked Plot

- uses bars to show comparisons between categories of data, but with the ability to break down and compare parts as a whole

```python
df.plot(kind = 'bar', stacked = True, title = 'Samples')
```

# Data Visualization (Histogram)

### Data

```python
iris
```

### Histogram

- primary use is to display the distribution (or shape) of the values in a data series
- allows you to see frequency distribution of a data set

```python
iris.plot(kind = 'hist', title = 'Iris Dataset', bins = 50)
```

```python
# Stacked Histogram

iris.plot(kind = 'hist', title = 'Iris Dataset', bins = 50, stacked = True)
```

```python
# Orientation

iris.plot(kind = 'hist', title = 'Iris Dataset', bins = 50, stacked = True, orientation = 'horizontal')
```

```python
data
```

```python
# Alternate Histogram (separated for each category)

data.hist(color = 'r')
```

# Data Visualization (Box Plot)

```python
data
```

### Box Plot

- a box or whisker plot is a graph that presents information from a five number summary
- minimum, 1st quartile, median, 3rd quartile, maximum
- it summarizes variation in large datasets visually and shows outliers (minimum Q1 - 1.5 * IQR, maximum (Q3 + 1.5 * IQR)

```python
data.plot(kind = 'box', figsize = (15,5))
```

```python
data.describe()
```

```python
# Orientation

data.plot(kind = 'box', figsize = (15,5), vert = False)
```

# Data Visualization (Area Plot)

```python
# Data

data
```

### Area Plot

- an area plot combines the line chart and bar chart to show how one or more group's numeric values change over the progression of a second variable (distinguished by the addition of shading between lines)

```python
data.plot(kind = 'area', figsize = (10,5), title = 'Iris Dataset')
```

```python
# Non-Stacked, transparency

data.plot(kind = 'area', figsize = (10,5), stacked = False, alpha = 0.25, title = 'Iris Dataset')
```

# Data Visualization (Scatter Plot)

### Scatter Plot

- a scatter plot uses dots to represent values for two different numeric variables
- good from trying to show relationship between two variables

```python
# Data

data
```

```python
# x and y axis values

data.plot(x = 'sepal_width', y = 'petal_width', kind = 'scatter', title = 'Iris Dataset', c = 'sepal_width')
```

```python
data.plot(kind = 'scatter', x = 'sepal_length', y = 'petal_length', label = 'length')
```

```python
# Secondary Axis

ax = data.plot(x = 'sepal_width', y = 'petal_width', kind = 'scatter', label = 'width')
data.plot(kind = 'scatter', x = 'sepal_length', y = 'petal_length', ax = ax, color = 'r', label = 'length')
```

# Data Visualization (Pandas - Hex Plot)

```python
# Data

data
```

### Hex Plot

- hexagonal plots solves the problem of having too many points that overlap (it plots density, rather than points)
- hexagons are more similar to circle than a square (this translates to a more efficient data aggregation)

```python
data.plot(kind = 'hexbin', x = 'sepal_length', y = 'petal_length', gridsize = 20, C = 'sepal_width')
```

# Data Visualization (Pandas - Scatter Matrix)

```python
# Data

data
```

```python
# Import scatter matrix

from pandas.plotting import scatter_matrix
```

### Scatter Matrix

- is a grid (or matrix) of scatter plots used to visualize bivariate relationships between combinations of variables
- each scatter plot in the matrix visualizes the relationship between a pair of variables, allowing many relationship to be explored in one chart

```python
# Measures concentration values
# KDE (Kernel Density Estimation - creates a smooth curve to visualize the shape of the data)

scatter_matrix(data, figsize = (10,5), diagonal = 'kde', color = 'b') # kernel density estimate
```

```python
# Measures concentration values

scatter_matrix(data, figsize = (10,5), diagonal = 'hist', color = 'b') # histogram
```

# K - Means Clustering - 1

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Creating Artificial Data

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 101)

# data[0]
# data[0][:,0]
# data[0][:,1]
```

```python
data[0].shape
```

## Plots

```python
plt.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'rainbow') # Plotting column 1 and column 2
plt.grid()
```

## Clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4) # is unsupervised learning algorithms, we find labels into clusters
```

```python
kmeans.fit(data[0]) # fit to features
```

```python
kmeans.cluster_centers_ # centroid of clusters
```

```python
kmeans.labels_ # shows to where a data point belongs to a cluster (find labels)
```

```python
data[1] # original labels
```

## Plot Comparisons

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey = True, figsize = (15,6))

ax1.set_title('K - Means')
ax1.scatter(data[0][:,0], data[0][:,1], c = kmeans.labels_, cmap = 'rainbow')
ax1.grid()

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'rainbow')
ax2.grid()
```

# Data Visualization (MathPlotLib - Part 1)

```python
import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt # alternative
%matplotlib inline
```

### Generate Data

```python
# x-axis

x = np.linspace(1,20,20)
x
```

```python
np.random.seed(10)

# y-axis

y = np.random.randint(1,50,20)
y
```

```python
# Use plt

plt.plot(y)
```

```python
# Cumulative

y = y.cumsum()
```

```python
# Plot again

plt.plot(y)
```

## Line Plot

```python
# Plot x and y

plt.plot(x, y)
```

### Change Color and Style

```python
# Color (constant or hex values), Style (:, -., --, -)

plt.plot(x, y, color = 'r', marker = 'o', linestyle = '--')
```

### Sizing, Labels, Title

```python
plt.figure(figsize=(10,5))
plt.plot(x, y, color = 'r', marker = 'x', linestyle = '-')
plt.xlabel('X - axis', fontsize = 12)
plt.ylabel('Y - axis', fontsize = 12)
plt.title('Synthetic Graph', fontsize = 14)
```

## Scatter, Bar, Hist, Box Plots

### Scatter

```python
plt.figure(figsize=(10,5))
plt.scatter(x, y, marker = 's', color = 'g', linewidth = 3)
```

## Bar

```python
# Data

b = [10, 23, 43, 5, 66]
a = ['a', 'b', 'c', 'd', 'e']
```

```python
plt.bar(a, b, width = 0.75)
```

```python
plt.barh(a, b, height = 0.75)
```

## Histogram

```python
plt.hist(y, rwidth = 0.75, bins = 20)
```

## Box Plot

```python
# Data

data = [np.random.normal(0, std, 100) for std in range(1, 3)]
```

```python
plt.boxplot(data, vert = True, patch_artist = True)
plt.show()
```

## Subplots

```python
# Data

x
```

```python
y
```

```python
y2 = x * y
```

```python
y2
```

```python
plt.plot(x, y)
plt.plot(x, y2)
```

```python
# Plot lines separately

plt.figure(figsize = (10,5))

plt.subplot(1, 2, 1)
plt.plot(x, y, marker = 'o', color = 'r', markersize = 8)

plt.subplot(1, 2, 2)
plt.plot(x, y2, marker = '*', color = 'k', markersize = 8)
```

```python
# Nested Plots (Advanced)


fig = plt.figure(figsize = (12,5))
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.12, 0.6, 0.4, 0.3]) # top, bottom, (sizes)

ax1.plot(x, y, 'r')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_title('Y Plot')

ax2.plot(x, y, 'g')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.set_title('Y Plot')
```

### xlim, ylim, xticks, yticks

```python
fig, ax = plt.subplots(1, 2, figsize = (12,4))

ax[0].plot(x, y, x, y2)
ax[1].plot(x, y2, 'k')

ax[0].set_ylim([0, 1500])
ax[0].set_xlim([0, 50])

y_limits = np.arange(0,10100, 1000)

ax[1].set_xticks([1, 10, 5])
ax[1].set_yticks(y_limits.tolist())
plt.show()
```

### Pie Plot

```python
# Pie (OOP)
# Data

data = [3,8,10,12]
labels = ['Cat', 'Dog', 'Hamster', 'Rabbit']
```

```python
fig, ax = plt.subplots(figsize = (21,6), dpi = 110)

wedges, texts, autotexts = ax.pie(data, labels = labels, autopct = '%1.1f%%', explode = (0.2, 0.1, 0.1, 0), 
       shadow = True, startangle = 80, radius = 0.95, colors = ['r','b','y','g'], wedgeprops = {'linewidth': 2},
       labeldistance = 1.15)

ax.set_title('Pie Plot\n\n', fontsize = 15, weight = 'bold') # set title

plt.setp(autotexts, fontsize = 9, weight = 'bold', color = '#ffffff')
plt.setp(texts, fontsize = 10, weight = 'bold', color = 'b')
plt.setp(wedges, alpha = 0.75, edgecolor = 'k', linestyle = '--')

# ax.legend(wedges, labels, title = 'Pets', loc = 'upper right', fontsize = 8, borderpad = 1) # if you want legends

plt.show()
```

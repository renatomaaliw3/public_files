# Data Visualizations

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Sample Data

```python
# Sample Data

x = np.linspace(0, 10, 25)

y = np.cumsum(x)

y1 = np.sin(x)
y2 = np.cos(x)
```

## Plots

```python
# Setting Figure Size

plt.figure(figsize = (8,3))

# Line Graph

plt.plot(x, y1)
plt.plot(x, y2)
```

```python
# Customizations

# color (str) --> code or hex values (similar to CSS)
# linestyle (str) --> '-', '--', ':', '-.'
# linewidth (int)
# marker (str) --> . | , | o | v | ^ | < | > | 1 | 2 | 3 | 4 | s | p | * | h | H | + | x | D | d
# markersize (int)

plt.plot(x, y1, color = 'r', linestyle = '-', linewidth = 2, marker = 'p', markersize = 10, label = 'Sin(x)')
plt.plot(x, y2, color = 'b', linestyle = '--', linewidth = 2, marker = '*', markersize = 10, label = 'Cos(x)')

# Customize axes labels and title

plt.xlabel('X-Axis Label', fontsize = 13, color = 'orange', fontweight = 'bold', labelpad = 15)
plt.ylabel('Y-Axis Label', fontsize = 14, color = 'green', fontstyle = 'italic', labelpad = 15)
plt.title('Customized Line Graph', fontsize = 15, color = 'brown', pad = 15)

# Customize Ticks
# ha is horizontal alignment (left, right, center)
# va is vertical alignment (top, bottom, center, baseline, center_baseline)

plt.xticks(np.arange(0, 11, 1.5), rotation = 90, fontweight = 'bold', ha = 'center')
plt.yticks(np.arange(-2, 2, 0.5), rotation = 30, fontstyle = 'italic', va = 'center')

# Grids
plt.grid(True, linestyle = ':', alpha = 0.4, color = 'black') # which = 'minor' grid

# Add Legend
# loc is combinationo of center|upper|lower| left|right or tuple of (x, y)
plt.legend(loc = 'upper right', borderpad = 1, handletextpad = 1, facecolor = 'green', edgecolor = 'black') # bbox_to_anchor = (1,1) -> custom positions

# Customize the Plot Background

plt.gca().set_facecolor('#def123')

# Show

plt.show()
```

### Scatter

```python
# Generate random data

np.random.seed(42)
x3 = np.random.rand(50)
y3 = 2 * x3 + 1 + 0.1 * np.random.randn(50)
```

```python
plt.scatter(x3, y3, color = 'blue', marker ='o', s = 20, label='Data Points')

# Customize the plot

plt.title('Customized Scatter Plot')
plt.xlabel('X-axis Label', fontsize=12, color='purple', fontweight='bold')
plt.ylabel('Y-axis Label', fontsize=12, color='orange', fontstyle='italic')
plt.legend()
plt.show()
```

### Bar Graphs

```python
# Sample data

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 24, 30, 20]

plt.figure(figsize = (5,4))

# Create a bar graph with customizations

plt.bar(categories, values, color='skyblue', edgecolor='black', linewidth=2, alpha=0.8, width=0.6, label='Bar Chart') #barh --> height

# Customize the plot

plt.title('Customized Bar Graph')
plt.grid(alpha = 0.4, axis = 'y')
plt.xlabel('Categories', fontsize=12, color='purple', fontweight='bold')
plt.ylabel('Values', fontsize=12, color='orange', fontstyle='italic')

plt.legend()
```

### Histogram

```python
# Generate random data for the histogram

np.random.seed(42)
data = np.random.randn(1000)

# Create a histogram with customizations
plt.hist(data, bins=20, color='skyblue', edgecolor='black', linewidth=1.2, alpha=0.8)

# Customize the plot
plt.title('Customized Histogram')
plt.xlabel('Values', fontsize=12, color='purple', fontweight='bold')
plt.ylabel('Frequency', fontsize=12, color='orange', fontstyle='italic')
plt.show()
```

### Box Plot

```python
# Generate random data for the boxplot
np.random.seed(42)
data = [np.random.normal(0, 1, 100), np.random.normal(2, 1, 100), np.random.normal(3, 2, 100)]

# Boxplot with customizations
plt.boxplot(data, labels=['Group A', 'Group B', 'Group C'], patch_artist=True,
            boxprops=dict(facecolor='skyblue', edgecolor='black', linewidth=1.2),
            medianprops=dict(color='black', linewidth=2),
            whiskerprops=dict(color='green', linestyle='--'),
            capprops=dict(color='red')) # --> remember vert = False, for horizontal

# Customize the plot
plt.title('Customized Boxplot')
plt.xlabel('Groups', fontsize=12, color='purple', fontweight='bold')
plt.ylabel('Values', fontsize=12, color='orange', fontstyle='italic')

plt.show()
```

### Pie Chart

```python
# Sample data for the pie chart

labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 20, 25]
colors = ['skyblue', 'lightgreen', 'lightcoral', 'yellow']

# Create a pie chart with customizations
plt.pie(sizes, labels = labels, colors = colors, explode = (0.1, 0.2, 0.3, 0.3), autopct='%1.1f%%', shadow = True, startangle=65,
        textprops = {'color': 'black', 'fontsize': 12, 'fontweight': 'bold'}, pctdistance = 0.50, labeldistance = 1.2)

# Customize the plot
plt.title('Customized Pie Chart', pad = 15)
plt.show()
```

### Subplots

```python
# Generate sample data for the line graphs

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots with customizations

plt.figure(figsize=(10, 5))

plt.suptitle('Subplots')

# Subplot 1

plt.subplot(1, 2, 1)
plt.plot(x, y1, label='Sin(x)', color='blue', linestyle='-', linewidth=2, marker='o', markersize=8)
plt.title('Line Graph 1')
plt.xlabel('X-axis', fontsize=12, color='purple', fontweight='bold')
plt.ylabel('Y-axis', fontsize=12, color='orange', fontstyle='italic')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Subplot 2

plt.subplot(1, 2, 2)
plt.plot(x, y2, label='Cos(x)', color='green', linestyle='--', linewidth=2, marker='s', markersize=8)
plt.title('Line Graph 2')
plt.xlabel('X-axis', fontsize=12, color='purple', fontweight='bold')
plt.ylabel('Y-axis', fontsize=12, color='orange', fontstyle='italic')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Inset Graphs

```python
# Generate sample data for the main graph
x_main = np.linspace(0, 10, 100)
y_main = np.sin(x_main)

# Generate sample data for the inset graph
x_inset = np.linspace(2, 4, 50)
y_inset = np.sin(x_inset)

# Create the main graph
plt.figure(figsize=(10, 8))

# Main graph

plt.plot(x_main, y_main, label='Main Graph', color='blue', linestyle='-', linewidth=2)
plt.title('Main Graph')
plt.xlabel('X-axis (Main)', fontsize=12, color='purple', fontweight='bold')
plt.ylabel('Y-axis (Main)', fontsize=12, color='orange', fontstyle='italic')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Inset graph

left, bottom, width, height = [0.635, 0.19, 0.20, 0.20]
ax_inset = plt.axes([left, bottom, width, height])
ax_inset.plot(x_inset, y_inset, label='Inset Graph', color='red', linestyle='--', linewidth=1)
ax_inset.set_title('Inset Graph')
ax_inset.set_xlabel('X-axis (Inset)', fontsize=8, color='purple', fontweight='bold')
ax_inset.set_ylabel('Y-axis (Inset)', fontsize=8, color='orange', fontstyle='italic')
ax_inset.grid(True, linestyle='--', alpha=0.7)
ax_inset.legend()

# Show the plot
plt.show()
```

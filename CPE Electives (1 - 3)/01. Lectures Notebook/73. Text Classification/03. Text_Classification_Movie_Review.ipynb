```python
# Imports

import numpy as np
import pandas as pd
```

## Load a dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/NLP/moviereviews.tsv', sep = '\t')
df
```

```python
# Number of reviews

len(df)
```

```python
# Sample reviews

print(df['review'][0])
```

```python
# Sample reviews

print(df['review'][1])
```

```python
# Check for missing values

df.isnull().sum()
```

```python
# Remove NaN reviews

df.dropna(inplace = True)
```

```python
# Check again

df.isnull().sum()
```

```python
len(df)
```

```python
# How about blank reviews?

blanks = []

for i, label, reviews in df.itertuples():

  if reviews.isspace():

    blanks.append(i)

blanks
```

```python
# Remove blanks based from index

df.drop(blanks, inplace = True)
len(df)
```

```python
# Train-Test Split

from sklearn.model_selection import train_test_split

X = df['review']
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Pipeline for Vectorization

from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC

text_clf = Pipeline([('tfidf', TfidfVectorizer()),('clf', LinearSVC())])
text_clf.fit(X_train, y_train)
```

```python
# Predictions

predictions = text_clf.predict(X_test)
predictions
```

```python
# Metrics

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

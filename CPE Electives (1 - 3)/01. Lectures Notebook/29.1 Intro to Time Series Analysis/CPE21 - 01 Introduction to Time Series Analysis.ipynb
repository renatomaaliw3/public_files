```python
# Libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset

df = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/bitcoin_price.csv')
df.head()
```

## Time Series Indexing (Master this part)

```python
# Convert "Date" to datetime object (YYYY-MM-DD) or "%Y-%m-%d"

df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d')
df.head()
```

```python
# Set Date as index

df.set_index('Date', inplace = True)
df.tail()
```

```python
# Parsing Dates and Setting Index

df_02 = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/bitcoin_price.csv', parse_dates = True, index_col = 'Date')
df_02.head()
```

```python
# Select a specific day in our df

df_02.loc['2023-02-01']
```

```python
# Resample the data to a different time granularity

df_02.resample('QE').mean()
```

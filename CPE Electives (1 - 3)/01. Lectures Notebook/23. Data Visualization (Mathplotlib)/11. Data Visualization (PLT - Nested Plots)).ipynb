```python
# Imports

import numpy as np # numpy
import pandas as pd # pandas
import matplotlib.pyplot as plt # matplotlib
import seaborn as sns # seaborn

import pdb # python debugger
```

##**Subplotting**

Allows you to create multiple plots within a single figure, making it easier to compare data and visualize multiple charts side-by-side.

### Nested Plots

```python
# Generate some example data

x = np.linspace(1,20,20)
y = np.random.randint(1,50,20)
y = y.cumsum()
```

```python
fig = plt.figure(figsize = (12,5))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.09, 0.6, 0.3, 0.3]) # top, bottom, (sizes)
ax3 = fig.add_axes([0.65, 0.15, 0.3, 0.3]) # top, bottom, (sizes)

ax1.plot(x, y, color = 'red', linewidth = 3, marker = 'o', linestyle = '--', markersize = 8)
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_title('Main Plot')

ax2.bar(x, y, label = 'Sample Legend')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.set_title('Bar Chart')
ax2.grid(alpha = 0.25)
ax2.legend()

ax3.scatter(x, y, color = 'green', marker = '*', label = 'Sample Legend')
ax3.set_xlabel('X axis')
ax3.set_ylabel('Y axis')
ax3.set_title('Scatter Plot')
ax3.grid(alpha = 0.25)
ax3.legend()
```

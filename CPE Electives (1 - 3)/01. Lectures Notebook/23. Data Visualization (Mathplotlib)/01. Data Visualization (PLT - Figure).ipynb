# Data Visualizations (Part - 01)

```python
# Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Introduction to 'plt.figure'

  -- is a core function in Matplotlib, used to create and configure a new figure for plotting.
  
  -- a "figure" in Matplotlib is an object that holds all elements of a plot, including axes, titles, labels, and other visual elements.

```python
# A. Figure (figure)

plt.figure()
plt.plot([1, 2, 3], [4, 5, 6]) # This is a basic line graph
plt.show()
```

```python
# B. (figsize)
# You can control the width and height of the figure in inches via figure size

plt.figure(figsize = (10, 3)) # 10 (width) 3 (height) tall
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# C. DPI (Dots Per Inch)

plt.figure(figsize = (10, 5), dpi = 100) # 100 DPI for moderate resolution
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# D. Background Color (facecolor)

plt.figure(figsize = (10,5), facecolor = '#abc123') # red background
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# E. Frame Visibility (frameon)

plt.figure(figsize = (10,5), facecolor = 'yellow', frameon = False) # Removes the border frame
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# F. Figure Title (frameon)
# While not a direct parameter of plt.figure, the title can be added to the figure
# using suptitle

fig = plt.figure()
fig.suptitle('Main Figure Title')
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# G. Saving a Figure (savefig)
# While not a direct parameter of plt.figure, the title can be added to the figure
# using suptitle

fig = plt.figure(figsize = (10, 5), facecolor = 'red', dpi = 100)
fig.suptitle('Main Figure Title')
plt.plot([1, 2, 3], [4, 5, 6])
fig.savefig('my_figure.png')
```

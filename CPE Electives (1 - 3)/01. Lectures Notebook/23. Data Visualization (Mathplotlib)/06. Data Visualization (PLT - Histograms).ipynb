#**Histograms**

A histogram is a graphical representation of the distribution of numerical data. It is widely used in statistics and data analysis to visualize the frequency or density of data within specified intervals, known as "bins." Each bin represents a range of values, and the height (or length) of each bar indicates the number of data points (or frequency) within that range.

##**Purpose of Histogram**:

  -- _**Visualize Data Distribution:**_: Histograms provide a quick visual representation of data distribution across intervals, helping to identify whether the data is symmetrical, skewed, or has a particular shape.

  -- _**Analyze Frequency and Density**_: By displaying the frequency or density of data within specified intervals, histograms reveal where data points are concentrated, as well as the overall spread of the data.

  -- _**Identify Patterns and Trends**_: Histograms make it easy to spot common patterns, such as normal, bimodal, uniform, or skewed distributions, which can inform further statistical analysis.

  -- _**Detect Outliers**_: Outliers or unusual values are easily identified in a histogram, as they often fall into sparse or empty bins on either end of the chart.

  -- _**Compare Datasets**_: Histograms allow for easy comparison of different datasets (e.g., overlayed or stacked histograms) to observe similarities or differences in distributions.

```python
# Dataset

df = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/histogram_data-02.csv')
df
```

```python
# A1. Basic Histogram

plt.figure(figsize = (12, 5))

# Histogram
plt.hist(df['normal_dist'])
plt.title("Basic Histogram (Normal Distribution)")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# A2. Basic Histogram (Orientation)

plt.figure(figsize = (12, 5))

# Histogram
plt.hist(df['normal_dist'], orientation = 'horizontal')
plt.title("Basic Histogram (Normal Distribution)")
plt.xlabel("X - Values")
plt.ylabel("Distribution")
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# B. Basic Histogram with specified bins

plt.figure(figsize = (12, 5))

# Histogram
plt.hist(df['normal_dist'], bins = 40, color = 'green')
plt.title("Basic Histogram (Normal Distribution)")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# C. Basic Histogram with Label

plt.figure(figsize = (12, 5))

# Histogram
plt.hist(df['normal_dist'], bins = 30, color = 'green', edgecolor = 'black', label = 'Normal Distribution')
plt.title("Basic Histogram (Normal Distribution)")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.legend(loc = 'upper left')
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# D. Basic Histogram with Label, Outline

plt.figure(figsize = (12, 5))

# Histogram
plt.hist(df['normal_dist'], bins = 30, color = 'green',
         edgecolor = 'black', histtype = 'step', label = 'Normal Distribution')

plt.title("Basic Histogram (Normal Distribution)")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.legend()
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# E. Basic Histogram (Cumulative)

plt.figure(figsize = (12, 5))

# Histogram
plt.hist(df['normal_dist'], bins = 30, color = 'green',
         edgecolor = 'black', histtype = 'step', cumulative = True, label = 'Normal Distribution')

plt.title("Basic Histogram (Normal Distribution) - Cumulative")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.legend()
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# F. Basic Histogram (Log)

plt.figure(figsize = (12, 5))

# Histogram
plt.hist(df['normal_dist'], bins = 30, color = 'green',
         edgecolor = 'black', histtype = 'step', log = True, label = 'Normal Distribution')

plt.title("Basic Histogram (Normal Distribution) - Log")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.legend()
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# E. Histogram with KDE (Seaborn)

plt.figure(figsize = (12, 5))

# Histogram
sns.histplot(df['normal_dist'], bins = 30, color = 'green', edgecolor = 'black', label = 'Distribution', kde = True)
plt.title("Basic Histogram (Normal Distribution) and KDE")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.legend()
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# F1. Non-Stacked Histogram

plt.figure(figsize = (12, 5))

# Histogram
plt.hist([df['normal_dist'], df['uniform_dist']], bins = 50, color = ['green', 'red'], edgecolor = 'black', label = ['Normal Distribution', 'Uniform Distribution'])

plt.title("Non-Stacked Histogram")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.legend()
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# F2. Stacked Histogram

plt.figure(figsize = (8, 5))

# Histogram
plt.hist([df['normal_dist'], df['exponential_dist'], df['positively_skewed']],
         bins = 40, stacked = True, color = ['green', 'red', 'blue'],
         edgecolor = 'black', label = ['Normal Distribution', 'Exponential Distribution', 'Positively Skewed'])

plt.title("Stacked Histogram")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.legend()
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

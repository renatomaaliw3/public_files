#**Pie Chart**

A Pie Chart is a circular statistical graphic used to illustrate the proportions of a whole. Each segment (or "slice") of the pie represents a percentage or part of the total, with the entire chart equaling 100%

##**Purpose of Pie Chart**:

  -- _**Show Part-to-Whole Relationships**_: Pie charts are ideal for illustrating how individual categories contribute to a larger whole. This makes them useful for demonstrating proportions, distributions, and contributions.

  -- _**Provide a Quick Visual Summary**_: Pie charts simplify data, allowing viewers to quickly grasp the general composition or distribution of a dataset without needing to dive into numbers. This is particularly effective for high-level or non-technical audiences..

  -- _**Enhance Comparative Analysis**_: While pie charts aren’t ideal for precise comparisons, they work well for showcasing the relative size of categories. For example, it’s easy to see which category is largest or smallest in relation to others.

  -- _**Comparing Groups or Categories**_: Some pie charts use techniques like "exploding" slices (moving a slice outward) to highlight certain categories, which can emphasize particular points in presentations or reports.

  -- _**Emphasize Significant Categories:**_: Scatter plots offer an intuitive view of the data distribution, which is helpful for assessing clustering and spread across the range of the data values.

```python
# A. Basic Pie Chart

plt.figure(figsize = (5,5))

# Data to plot
labels = ['Apples', 'Bananas', 'Cherries', 'Dates'] # assigned labels
sizes = [15, 30, 45, 10] # defines the size of the wedge

# Plot
plt.pie(sizes, labels = labels) # default settings of pie chart
plt.title('Basic Pie Chart') # title
plt.show()
```

```python
# B. Pie Chart with Custom Colors

plt.figure(figsize = (5,5))

# Data to plot
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Plot
plt.pie(sizes, labels = labels, colors = colors)
plt.title('Pie Chart with Custom Colors')
plt.show()
```

```python
# C. Exploded Pie Chart

plt.figure(figsize = (5,5))

# Data to plot
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]
explode = (0.3, 0.15, 0.1, 0.2)  # Explode values

# Plot
plt.pie(sizes, labels = labels, explode = explode)
plt.title('Exploded Pie Chart', pad = 20)
plt.show()
```

```python
# D. Pie Chart with Percentage Labels

plt.figure(figsize = (5,5))

# Data to plot
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]

# Plot
plt.pie(sizes, labels = labels, autopct = '%.0f%%')
plt.title('Pie Chart with Percentages')
plt.show()
```

```python
# E. Pie Chart with Shadow and Start Angle

plt.figure(figsize = (5,5))

# Data to plot
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]
explode = (0.1, 0, 0, 0)

# Plot
plt.pie(sizes, labels = labels, shadow = True, explode = explode,
        autopct = '%.2f%%', startangle = 270)

plt.title('Pie Chart with Shadow and Start Angle')
plt.show()
```

```python
# F1. Pie Chart with Border Colors & Line Style, etc.

plt.figure(figsize = (5,5))

# Data to plot
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]
colors = ['red', 'green', 'orange', 'yellow']
explode = (0, 0, 0, 0)

# Plot
plt.pie(sizes, labels = labels, colors =  colors, explode = explode,
        autopct = '%.0f%%', startangle = 190, pctdistance = 0.5,
        wedgeprops = dict(edgecolor = 'black', ls = '-', lw = 3,
        hatch = '', alpha = 1, fill = True)) # has also a width property

plt.title('Pie Chart with Customizations')
plt.show()
```

```python
# F2. Pie Chart with Border Colors & Line Style, etc.

plt.figure(figsize = (5,5))

# Data to plot
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]
colors = ['red', 'green', 'orange', 'yellow']
explode = (0.1, 0.05, 0.05, 0.05)

# Plot
wedges, texts, autotexts = plt.pie(sizes, labels = labels, colors =  colors, explode = explode, autopct = '%.0f%%',
                                    startangle = 120, counterclock = False, shadow = True,
                                    wedgeprops = dict(edgecolor = 'black', ls = '--', lw = 2),
                                    textprops = dict(color = 'black', fontsize = 11, rotation = 0, fontweight = 'bold'),
                                    labeldistance = 1.1, pctdistance = 0.75,
                                    hatch = '')

# Set font size for labels only
for text in texts:

    text.set_fontsize(12)  # Set the font size you want for labels
    text.set_color('red')

plt.title('Pie Chart with Customizations', pad = 20)
plt.show()
```

```python
# G. Pie Chart with Legend

plt.figure(figsize = (6, 6))

# Data to plot
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]
explode = (0.1, 0.05, 0.05, 0.05)

# Plot
patches, texts, autotexts = plt.pie(sizes, explode = explode, shadow = True,
                                    autopct = '%.0f%%', startangle = 190,
                                    wedgeprops = dict(edgecolor = 'black'),
                                    textprops = dict(fontsize = 14, color = 'white'))

# Legend
plt.legend(patches, labels, loc = 'upper left', fontsize = 7)

plt.axis('equal') # Ensure the pie chart is drawn as a circle

plt.title('Pie Chart with Legend')
plt.show()
```

```python
# Sample Data

data = pd.read_csv("https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/nba%20cleaned.csv")
data
```

```python
# Create a Pie Chart based on the proportions of unique positions in the league (NBA)

positions = data.value_counts('Position')
positions
```

```python
import math

plt.figure(figsize = (10, 10))

# Data to plot
labels = list(positions.index)
sizes = list(positions)

def format_autopct(pct, allvals):

  absolute = int(pct/99.99 * sum(allvals)) # If discrepancies, adjust to 99.99 or 99.98, etc. for int(), or use math.ceil if possible at 100.10
  # absolute = math.ceil(pct/100.10 * sum(allvals)) # If discrepancies, adjust to 99.99 or 99.98, etc. for int(), or use math.ceil if possible at 100.10
  return f"{absolute} ({pct:.2f}%)"

# Plot
plt.pie(sizes, labels = labels,
        autopct = lambda pct: format_autopct(pct, sizes), startangle = 190, explode = [0.05] * len(positions.index),
        wedgeprops = dict(edgecolor = 'black', lw = 1, fill = True),
        textprops = dict(fontsize = 9)) # has also a width property


plt.title('NBA Positions')
plt.show()
```

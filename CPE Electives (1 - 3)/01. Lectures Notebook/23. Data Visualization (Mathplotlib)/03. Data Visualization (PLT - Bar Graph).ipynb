#**Bar Graphs**

Bar graphs are one of the most commonly used types of charts in data visualization, primarily because of their simplicity and effectiveness in representing categorical data. Here is a comprehensive overview of the purpose and uses of bar graphs:

##**Purpose of Bar Graphs**:

  -- _**Compare Values Across Categories**_: They allow easy comparison of data across discrete categories, showing the magnitude of each category clearly.

  -- _**Highlight Differences**_: Bar graphs make it easy to spot the highest and lowest values among categories, as each barâ€™s length directly corresponds to its value.

  -- _**Show Distribution of Data**_: In certain contexts, such as histograms (a specific type of bar graph), they represent the distribution of a variable by grouping data into intervals (bins).

  -- _**Track Changes Over Time**_: Though line graphs are more common for time-series data, bar graphs can also represent changes over time, especially when comparing discrete time intervals (e.g., annual sales figures).

```python
# A1. Simple Bar Graphs

# Data for the bar graph
categories = ['Store 1', 'Store 2', 'Store 3', 'Store 4']
values = [3, 7, 5, 2]

# Create a basic bar graph
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color = 'red')

# Add labels and title
plt.xlabel('Stores')
plt.ylabel('No. Sold')
plt.title('Company X')

# Show grid
plt.grid(True, alpha = 0.5, color = 'gray')

plt.show()
```

```python
# A2. Simple Bar Graphs (hatch, edgecolor, color, width)

# Data for the bar graph
categories = ['Store 1', 'Store 2', 'Store 3', 'Store 4']
values = [3, 7, 5, 2]

# Create a basic bar graph
plt.figure(figsize=(6, 4))
plt.bar(categories, values, hatch = '/x-', edgecolor = 'lightgreen', color = 'black', width = 0.5)  # you can set specific width, hatch

# Add labels and title
plt.xlabel('Stores')
plt.ylabel('No. Sold')
plt.title('Company X')

# Show grid
plt.grid(True, alpha = 0.5, color = 'gray')

plt.show()

# Possible hatch values

# "/", "\", "|", "-", "+", "x", "o", "O", ".", and "*".
```

```python
# A3. Simple Bar Graphs (with error bars)

# Data for the bar graph
categories = ['Store 1', 'Store 2', 'Store 3', 'Store 4']
values = [3, 7, 5, 2]

# Create a basic bar graph with error bars
errors = [0.25, 0.35, 0.45, 0.5]
plt.figure(figsize=(16, 4))
plt.bar(categories, values, yerr = errors, capsize = 0)  # error bars (yerr, xerr), capsize = float

# Add labels and title
plt.xlabel('Stores')
plt.ylabel('No. of Sold')
plt.title('Company X')

# Show grid
plt.grid(True, alpha = 0.25, color = 'gray')

plt.show()
```

```python
# B. Horizontal Bar Graphs

# Data for the bar graph
categories = ['Store 1', 'Store 2', 'Store 3', 'Store 4']
values = [3, 7, 5, 2]

# Create a basic bar graph (horizontal)
plt.figure(figsize=(6, 4))
plt.barh(categories, values) # you can set specific height

# Add labels and title
plt.xlabel('No. of Sold')
plt.ylabel('Stores')
plt.title('Company X')

# Show grid
plt.grid(True, alpha = 0.5, color = 'gray')

plt.show()
```

## Grouped Bar Graph

```python
# Grouped Bar Graph
# A grouped bar graph shows multiple series of data side-by-side

# Data for the bar graph
categories = ['Bags', 'Pants', 'Watch', 'Shirts']
values = [3, 7, 5, 2]
values2 = [4, 6, 2, 5]

x = np.arange(len(categories))

# Define the width of the bars
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, values, width, label='Male')
plt.bar(x + width/2, values2, width, label='Female')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('No. Sold')
plt.title('Store X')

# Ticks
plt.xticks(x, categories)

# Legend
plt.legend(loc = 'upper left')

# Grid
plt.grid(True, alpha = 0.5, linestyle = 'dotted', color = 'green')

plt.show()
```

```python
# Grouped Bar Graph (3 Groups)
# A grouped bar graph shows multiple series of data side-by-side

# Data for the bar graph
categories = ['Store 1', 'Store 2', 'Store 3', 'Store 4']
values = [3, 7, 5, 2]
values2 = [4, 6, 2, 5]
values3 = [5, 7, 3, 6]

x = np.arange(len(categories))

# Define the width of the bars
width = 0.25

plt.figure(figsize=(8, 5))
plt.bar(x - width, values, width, label ='Bags', color = 'red')        # Shift Left
plt.bar(x, values2, width, label = 'Pants', color = 'green')             # Centered
plt.bar(x + width, values3, width, label = 'Shirts', color = 'blue')      # Shift Right

# Add labels and title
plt.xlabel('Stores')
plt.ylabel('No. Sold')
plt.title('Company X')

# Ticks
plt.xticks(x, categories)

# Legend
plt.legend(loc = 'upper left')

# Grid
plt.grid(True, alpha = 0.5, linestyle = 'dotted', color = 'green')

plt.show()
```

```python
# Grouped Bar Graph (4 Groups)
# A grouped bar graph shows multiple series of data side-by-side

# Data for the bar graph
categories = ['Store 1', 'Store 2', 'Store 3', 'Store 4']
values =  [3, 7, 5, 2]
values2 = [4, 6, 2, 5]
values3 = [5, 7, 3, 6]
values4 = [2, 5, 4, 7]

x = np.arange(len(categories))

# Define the width of the bars
width = 0.20

plt.figure(figsize=(8, 5))
plt.bar(x - 1.5 * width, values, width, label='Bags')       # see grouped bar formula
plt.bar(x - 0.5 * width, values2, width, label='Pants')      # see grouped bar formula
plt.bar(x + 0.5 * width, values3, width, label='Watches')      # see grouped bar formula
plt.bar(x + 1.5 * width, values4, width, label='Shirts')      # see grouped bar formula

# Add labels and title
plt.xlabel('Stores')
plt.ylabel('No. Sold')
plt.title('Company X')

# Ticks
plt.xticks(x, categories)

# Legend
plt.legend(loc = 'upper left', fontsize = 9, borderaxespad = 1, borderpad = 0.75)

# Grid
plt.grid(True, alpha = 0.5, linestyle = 'dotted', color = 'green')

plt.show()
```

```python
# Grouped Bar Graph (4 Groups)
# A grouped bar graph shows multiple series of data side-by-side

# Data for the bar graph
categories = ['Store 1', 'Store 2', 'Store 3', 'Store 4']
values =  [3, 7, 5, 2]
values2 = [4, 6, 2, 5]
values3 = [5, 7, 3, 6]
values4 = [2, 5, 4, 7]

y = np.arange(len(categories))

# Define the width of the bars
width = 0.20

plt.figure(figsize=(8, 5))
plt.barh(y - 1.5 * width, values, width, label='Bags')       # see grouped bar formula
plt.barh(y - 0.5 * width, values2, width, label='Pants')      # see grouped bar formula
plt.barh(y + 0.5 * width, values3, width, label='Watches')      # see grouped bar formula
plt.barh(y + 1.5 * width, values4, width, label='Shirts')      # see grouped bar formula

# Add labels and title
plt.xlabel('No. Sold', labelpad = 10)
plt.ylabel('Stores', labelpad = 10)
plt.title('Company X', pad = 10)

# Ticks
plt.yticks(y, categories)

# Legend
plt.legend(loc = 'best', fontsize = 9,
           borderaxespad = 0.5,
           borderpad = 0.75,
           edgecolor = 'black',
           framealpha = 0.5,
           facecolor = 'white')

# Grid
plt.grid(True, alpha = 0.5, linestyle = 'dotted', color = 'green')

# Invert the y-axis to make Store 1 appear at the top
plt.gca().invert_yaxis()

plt.show()
```

## Simple Bar Graphs (with Annotated values)

```python
# A1-1. Simple Bar Graphs with Annotated Values (Top)

# Data for the bar graph
categories = ['Store 1', 'Store 2', 'Store 3', 'Store 4']
values = [3, 7, 5, 2]

# Create a basic bar graph
plt.figure(figsize=(6, 4))
bars = plt.bar(categories, values, color = 'orange')

# Add labels and title
plt.xlabel('Stores')
plt.ylabel('No. Sold')
plt.title('Company X')

# Add values on top of each bar

for bar in bars:

    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval - 0.75, yval, ha = 'center', va = 'bottom', color = 'blue')

# Show grid
plt.grid(True, alpha = 0.5, color = 'gray')

# Adjust ylim
plt.ylim(0, np.max(values) * 1.25)

plt.show()
```

```python
# A4. Stacked (Vertical, Annotated Values)

# Data
categories = ['Store 1', 'Store 2', 'Store 3', 'Store 4']
values1 = [3, 7, 5, 2]
values2 = [4, 6, 2, 5]
values3 = [2, 3, 4, 1]

# Plot stacked bar chart
plt.figure(figsize=(8, 5))

bars1 = plt.bar(categories, values1, label = 'Bags', linewidth = 20)
bars2 = plt.bar(categories, values2, bottom = values1, label = 'Pants')
bars3 = plt.bar(categories, values3, bottom = np.add(values1, values2), label = 'Shirts')

# Add labels and title
plt.xlabel('Stores')
plt.ylabel('No. of Sold')
plt.title('Company X')

# Add a legend with a background color
plt.legend(edgecolor = 'black', facecolor = 'lightgray')

# Helper function

def add_values(bars, values, bottoms):

  for bar, val, bottom in zip(bars, values, bottoms):

    plt.text(bar.get_x() + bar.get_width() / 2, bottom + val / 2, val, ha = 'center', va = 'center')

# Calculate cumulative bottoms for each segment

bottoms1 = [0] * len(values1)
bottoms2 = values1
bottoms3 = np.add(values1, values2)

# Add values at the center of each bar segment

add_values(bars1, values1, bottoms1)
add_values(bars2, values2, bottoms2)
add_values(bars3, values3, bottoms3)

# Grid
plt.grid(True, alpha = 0.5, linestyle = 'dotted', color = 'green')

plt.show()
```

```python
# A6. Stacked (Vertical, Annotated Values, Total at the Top)

# Data
categories = ['Store 1', 'Store 2', 'Store 3', 'Store 4']
values1 = [3, 7, 5, 2]
values2 = [4, 6, 2, 5]
values3 = [2, 3, 4, 1]

# Plot stacked bar chart
plt.figure(figsize=(8, 6))

bars1 = plt.bar(categories, values1, label = 'Bags', linewidth = 20)
bars2 = plt.bar(categories, values2, bottom = values1, label = 'Pants')
bars3 = plt.bar(categories, values3, bottom = np.add(values1, values2), label = 'Shirts')

# Add labels and title
plt.xlabel('Stores')
plt.ylabel('No. of Sold')
plt.title('Company X')

# Add a legend with a background color
plt.legend(edgecolor = 'black', facecolor = 'lightgray')

# Helper function

def add_values(bars, values, bottoms):

  for bar, val, bottom in zip(bars, values, bottoms):

    plt.text(bar.get_x() + bar.get_width() / 2, bottom + val / 2, val, ha = 'center', va = 'center', fontsize = 8)

# Calculate cumulative bottoms for each segment

bottoms1 = [0] * len(values1)
bottoms2 = values1
bottoms3 = np.add(values1, values2)

# Add values at the center of each bar segment

add_values(bars1, values1, bottoms1)
add_values(bars2, values2, bottoms2)
add_values(bars3, values3, bottoms3)

# Calculate and display totals at the top of each stacked bar

totals = np.add(values1, np.add(values2, values3))

for i, total in enumerate(totals):

  plt.text(i, total + 0.5, total, ha = 'center', va = 'bottom', fontsize = 13, fontweight = 'bold', color = 'blue')

# Set ylim

plt.ylim(0, np.max(totals) * 1.5)

# Grid
plt.grid(True, alpha = 0.5, linestyle = 'dotted', color = 'green')

plt.show()
print(totals)
```

```python
# A7. Stacked (Horizontal, Annotated Values, Total at the left)

# Data
categories = ['Store 1', 'Store 2', 'Store 3', 'Store 4']
values1 = [3, 7, 5, 2]
values2 = [4, 6, 2, 5]
values3 = [2, 3, 4, 1]

# Plot stacked horizontal bar chart
plt.figure(figsize=(8, 6))

bars1 = plt.barh(categories, values1, label='Bags', linewidth=20)
bars2 = plt.barh(categories, values2, left=values1, label='Pants')
bars3 = plt.barh(categories, values3, left=np.add(values1, values2), label='Shirts')

# Add labels and title
plt.ylabel('Stores')
plt.xlabel('No. of Sold')
plt.title('Company X')

# Add a legend with a background color
plt.legend(edgecolor='black', facecolor='lightgray')

# Helper function to add values at the center of each bar segment
def add_values(bars, values, lefts):

    for bar, val, left in zip(bars, values, lefts):

        plt.text(left + val / 2, bar.get_y() + bar.get_height() / 2, val, ha='center', va='center', fontsize=8)

# Calculate cumulative left positions for each segment
lefts1 = [0] * len(values1)
lefts2 = values1
lefts3 = np.add(values1, values2)

# Add values at the center of each bar segment
add_values(bars1, values1, lefts1)
add_values(bars2, values2, lefts2)
add_values(bars3, values3, lefts3)

# Calculate and display totals at the end of each stacked bar
totals = np.add(values1, np.add(values2, values3))

for i, total in enumerate(totals):

    plt.text(total + 0.5, i, total, ha='left', va='center', fontsize=13, fontweight='bold', color='blue')

# Set xlim to accommodate total labels
plt.xlim(0, np.max(totals) * 1.5)

# Grid
plt.grid(True, alpha=0.5, linestyle='dotted', color='green')

plt.show()
```

#**Scatter Plots**

A scatter plot is a type of data visualization that displays values for typically two numerical variables using Cartesian coordinates.

In a scatter plot, each observation in the dataset is represented as a single point, with its position determined by the values of the two variables: one variable is assigned to the x-axis, and the other is assigned to the y-axis.

##**Purpose of Scatter Plots**:

  -- _**Identifying Relationships Between Variables**_: Scatter plots are ideal for visualizing potential relationships between two numerical variables. For example, you might use a scatter plot to see if there is a correlation between a person's height and weight.

  -- _**Detecting Patterns & Trends**_: Scatter plots help identify patterns and trends in data, such as clusters of points or specific data patterns, like linear, exponential, or other types of trends.

  -- _**Spotting Outliers**_: Scatter plots are effective in identifying outliers, or points that differ significantly from other points in the dataset. Outliers can provide valuable insights or indicate errors in the data..

  -- _**Comparing Groups or Categories**_: When points are color-coded or use different shapes, scatter plots can compare different groups within the data. This can highlight differences or similarities across categories.

  -- _**Evaluating Data Distribution**_: Scatter plots offer an intuitive view of the data distribution, which is helpful for assessing clustering and spread across the range of the data values.

```python
# A. Basic Scatter Plot

# Create a basic scatter plot
plt.figure(figsize=(6, 4))

# Plot Variable 1 against Variable 4
plt.scatter(df['Variable 1'], df['Variable 4'])

# Add labels and title
plt.xlabel('V1', fontsize = 13, labelpad = 10)
plt.ylabel('V4', fontsize = 13, labelpad = 10)
plt.title('Basic Scatter Plot', pad = 10, fontweight = 'bold')

# Show grid
plt.grid(True, alpha = 0.25, color = 'gray')

plt.show()
```

```python
# B. Basic Scatter Plot (color, marker, marker shape, size, edgecolor, linewidths = 3)

# Create a basic scatter plot
plt.figure(figsize=(12, 4))

# Plot Variable 1 against Variable 4
plt.scatter(df['Variable 1'], df['Variable 4'], color = 'red',
                                                marker = 'h',
                                                s = 150,
                                                alpha = 0.5,
                                                edgecolor = 'black',
                                                linewidths = 2)

# Add labels and title
plt.xlabel('V1', fontsize = 13, labelpad = 10)
plt.ylabel('V4', fontsize = 13, labelpad = 10)
plt.title('Basic Scatter Plot', pad = 10, fontweight = 'bold')

# Show grid
plt.grid(True, alpha = 0.25, color = 'gray')

plt.show()

# Link to possible marker values: https://matplotlib.org/stable/api/markers_api.html
# Link to cmap values: https://matplotlib.org/stable/users/explain/colors/colormaps.html
```

```python
# B2. Basic Scatter Plot (color bar)

# Create a basic scatter plot
plt.figure(figsize=(12, 7))

# Plot Variable 1 against Variable 4 with Variable 1 as the color map
scatter = plt.scatter(df['Variable 1'], df['Variable 4'],
                      c=df['Variable 1'],  # Use Variable 1 for color mapping
                      cmap='Blues',      # Choose a color map
                      marker='o',
                      s=80,
                      alpha=0.75,
                      edgecolor='black',
                      linewidths=2)
# Add color bar

cbar = plt.colorbar(scatter, orientation = 'vertical',
                    fraction = 0.15, shrink = 0.75)

cbar.set_label('Variable 1', fontsize=10, labelpad = 10)

# Add labels and title
plt.xlabel('V1', fontsize=13, labelpad=10)
plt.ylabel('V4', fontsize=13, labelpad=10)
plt.title('Scatter Plot with Color Bar for Variable 1', pad=10, fontweight='bold')

# Show grid
plt.grid(True, alpha=0.25, color='gray')
plt.tight_layout()

plt.show()
```

```python
# H. Scatter Plot (Highlight Specific Data, Annotation using Arrows (shrinkA, shrinkB), Circle per Groups)

# Create a basic scatter plot
plt.figure(figsize=(12, 5))

# Sort DataFrame by 'Label' to ensure consistent plotting order
sorted_df = df.sort_values('Label')

# Define specific colors for each group
specific_colors = ['red', 'green', 'blue']  # Ensure it matches the number of unique labels

# Plot Variable 1 against Variable 4, with each group having a specific color
for idx, label in enumerate(sorted_df['Label'].unique()):

    subset = sorted_df[sorted_df['Label'] == label]
    plt.scatter(subset['Variable 1'], subset['Variable 4'], label=label,
                marker='o', s=50, alpha=0.5, edgecolor='black', color=specific_colors[idx])

    # Calculate the center and radius for the circle
    x_center, y_center = subset['Variable 1'].mean(), subset['Variable 4'].mean()
    radius = np.sqrt(((subset['Variable 1'] - x_center) ** 2 + (subset['Variable 4'] - y_center) ** 2).max())

    # Add a circle around each group
    circle = plt.Circle((x_center, y_center), radius, color=specific_colors[idx],
                        fill=False, linestyle='--', linewidth=1.5)

    plt.gca().add_patch(circle)

# Highlight specific points (rows 25, 50, and 100) with a black star marker
highlight_indices = [25, 50, 100]
plt.scatter(df['Variable 1'].iloc[highlight_indices], df['Variable 4'].iloc[highlight_indices],
            color = 'black', marker = '*', s = 150, label = 'Highlighted Points', alpha = 0.75, edgecolor = 'orange')

# Annotate the highlighted points with arrows
for idx in highlight_indices:

    x = df['Variable 1'].iloc[idx]
    y = df['Variable 4'].iloc[idx]

    plt.annotate(f'P{idx}', xy = (x, y), xytext=(x, y + 0.15),
                 fontsize=12, color='black', ha='center', va='top',
                 arrowprops = dict(facecolor='black',
                                   arrowstyle='simple',
                                   lw = 1, shrinkB = 7, edgecolor = 'yellow'))

# Add labels and title
plt.xlabel('V1', fontsize=13, labelpad=10)
plt.ylabel('V4', fontsize=13, labelpad=10)
plt.title('Scatter Plot with Dashed Circles Around Each Group', pad=10, fontweight='bold')

# Show Legend
plt.legend(loc='best')

# Show grid
plt.grid(True, alpha=0.25, color='gray')

plt.show()
```

```python
# H. Scatter Plot (Highlight Specific Data, Annotation using Arrows (shrinkA, shrinkB), ConvexHull per Groups)

from scipy.spatial import ConvexHull
from matplotlib import patheffects

# Create a basic scatter plot (With ConvexHull)
plt.figure(figsize=(12, 5))

# Sort DataFrame by 'Label' to ensure consistent plotting order
sorted_df = df.sort_values('Label')

# Define specific colors for each group
specific_colors = ['red', 'green', 'blue']  # Ensure it matches the number of unique labels

# Plot Variable 1 against Variable 4, with each group having a specific color
for idx, label in enumerate(sorted_df['Label'].unique()):

    subset = sorted_df[sorted_df['Label'] == label]
    plt.scatter(subset['Variable 1'], subset['Variable 4'], label=label,
                marker='o', s=50, alpha=0.5, edgecolor='black', color=specific_colors[idx])

    # Calculate the convex hull for the subset of points
    points = subset[['Variable 1', 'Variable 4']].values
    hull = ConvexHull(points)

    # Plot the convex hull as a dashed line
    for simplex in hull.simplices:

        plt.plot(points[simplex, 0], points[simplex, 1], linestyle='--', color = specific_colors[idx], linewidth=1.5)

# Highlight specific points (rows 25, 50, and 100) with a black star marker
highlight_indices = [25, 50, 100]
plt.scatter(df['Variable 1'].iloc[highlight_indices], df['Variable 4'].iloc[highlight_indices],
            color = 'black', marker = '*', s = 150, label = 'Highlighted Points', alpha = 0.75, edgecolor = 'orange', zorder = 100)

# Annotate the highlighted points with arrows
for idx in highlight_indices:

    x = df['Variable 1'].iloc[idx]
    y = df['Variable 4'].iloc[idx]

    annotation = plt.annotate(f'P{idx}', xy = (x, y), xytext=(x, y + 0.15),
                 fontsize=12, color='black', ha='center', va='top',
                 arrowprops = dict(facecolor='black',
                                   arrowstyle='simple',
                                   lw = 1, shrinkB = 7, edgecolor = 'yellow'))

    # Apply path effects to create a border
    annotation.set_path_effects([patheffects.withStroke(linewidth = 2, foreground="yellow")]) # path effects


# Add labels and title
plt.xlabel('V1', fontsize=13, labelpad=10)
plt.ylabel('V4', fontsize=13, labelpad=10)
plt.title('Scatter Plot with Convex Hulls Around Each Group', pad = 10, fontweight='bold')

# Show Legend
plt.legend(loc='best')

# Show grid
plt.grid(True, alpha=0.25, color='gray')

plt.show()
```

```python
# I. Scatter Matrix Plot (Pandas)

from pandas.plotting import scatter_matrix

# Define specific colors for each group in the order of the unique labels

df_numerical = df[['Variable 1', 'Variable 2', 'Variable 3', 'Variable 4']]
sorted_df = df.sort_values('Label')

specific_colors = ['red', 'green', 'blue']
unique_labels = sorted_df['Label'].unique()   #Get unique labels in order of appearance

# Assign colors based on label position
colors = df['Label'].apply(lambda x: specific_colors[list(unique_labels).index(x)])

# Create the scatter matrix with colored points
plt.figure(figsize=(12, 12))
scatter_matrix(df_numerical, alpha=0.8, figsize=(12, 12),
               diagonal='kde', color=colors) # 'hist', 'kde'

# Add a legend manually
for label, color in zip(unique_labels, specific_colors):

    plt.scatter([], [], color = color, label=label)  # Empty scatter for legend entries

plt.legend(loc="center left", bbox_to_anchor = (1.05, 2.5), markerscale = 2, borderpad = 0.75, labelspacing = 1)

plt.suptitle('Scatter Matrix with Colors Based on Groups', fontsize = 16, fontweight='bold', y = 0.925)
plt.show()
```

```python
# I. Scatter Matrix Plot (Fixing Ticks)

from pandas.plotting import scatter_matrix
from matplotlib.ticker import FuncFormatter

# Select only numerical columns for the scatter matrix
df_numerical = df[['Variable 1', 'Variable 2', 'Variable 3', 'Variable 4']]
sorted_df = df.sort_values('Label')

# Define specific colors for each group in the order of the unique labels
specific_colors = ['red', 'green', 'blue']
unique_labels = sorted_df['Label'].unique()  # Get unique labels in order of appearance

# Assign colors based on label position
colors = df['Label'].apply(lambda x: specific_colors[list(unique_labels).index(x)])

# Define formatter function to limit ticks to two decimal places
formatter = FuncFormatter(lambda x, _: f'{x:.2f}')

# Create the scatter matrix with colored points
plt.figure()
axes = scatter_matrix(df_numerical, alpha=0.8, figsize=(12, 12),
                      diagonal='kde', color=colors)

# Apply formatter to all axes in the scatter matrix
for ax in axes.flatten():

    # axis formatter
    ax.xaxis.set_major_formatter(formatter)
    ax.yaxis.set_major_formatter(formatter)

# Add a legend manually
for label, color in zip(unique_labels, specific_colors):

    plt.scatter([], [], color=color, label=label)  # Empty scatter for legend entries

plt.legend(loc="upper left", bbox_to_anchor=(1.05, 4.03), markerscale=2, borderpad=0.75, labelspacing=1)
plt.suptitle('Scatter Matrix with Colors Based on Groups', fontsize=16, fontweight='bold', y=0.915)

plt.show()
```

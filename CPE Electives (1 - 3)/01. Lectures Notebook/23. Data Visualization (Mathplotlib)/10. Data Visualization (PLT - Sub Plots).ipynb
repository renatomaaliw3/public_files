##**Subplotting**

Allows you to create multiple plots within a single figure, making it easier to compare data and visualize multiple charts side-by-side.

### Simple Grid Layout

The **subplot()** function allows you to create a grid of subplots, specifying the number of rows and columns, as well as the plotâ€™s position within the grid.

```python
# Create a 1x2 grid of subplots

plt.figure(figsize=(10, 4))
plt.suptitle('Figure Title')

# First subplot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.title('Sub Plot 1')
plt.xlabel('SP1 x-axis')
plt.ylabel('SP1 y-axis')

# Second subplot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.title('Sub Plot 2')
plt.xlabel('SP2 x-axis')
plt.ylabel('SP2 y-axis')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
# Create a 2x2 grid of subplots

plt.figure(figsize=(10, 8))
plt.suptitle('Figure Title')

# First subplot
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.title('Sub Plot 1')
plt.xlabel('SP1 x-axis')
plt.ylabel('SP1 y-axis')

# Second subplot
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot
plt.title('Sub Plot 2')
plt.xlabel('SP2 x-axis')
plt.ylabel('SP2 y-axis')

# Third subplot
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot
plt.title('Sub Plot 3')
plt.xlabel('SP3 x-axis')
plt.ylabel('SP3 y-axis')

# Fourth subplot
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot
plt.title('Sub Plot 4')
plt.xlabel('SP4 x-axis')
plt.ylabel('SP4 y-axis')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

## Flexible Grid Layout

**subplots()** is more flexible, as it allows you to create the entire grid at once and returns figure **(fig)** and axis **(ax)** objects that you can use for more customization.

```python
# Create a 2x1 grid of subplots

fig, ax = plt.subplots(2, 1, figsize=(6, 8), dpi = 100)  # 2 rows, 1 column

fig.suptitle('Figure Title')

# First subplot

ax[0].set_title("Subplot 1")
ax[0].set_title('Sub Plot 1')
ax[0].set_xlabel('SP1 x-axis')
ax[0].set_ylabel('SP1 y-axis')

# Second subplot
ax[1].set_title("Subplot 2")
ax[1].set_title('Sub Plot 2')
ax[1].set_xlabel('SP2 x-axis')
ax[1].set_ylabel('SP2 y-axis')

plt.tight_layout()
plt.show()
```

```python
# Create a 2x2 grid of subplots
fig, ax = plt.subplots(2, 2, figsize=(10, 8), dpi=100)  # 2 rows, 2 columns

# Add a main title for the entire figure
fig.suptitle('Figure Title', y = 1)

# First subplot
ax[0, 0].set_title("Subplot 1")
ax[0, 0].set_xlabel('SP1 x-axis')
ax[0, 0].set_ylabel('SP1 y-axis')

# Second subplot
ax[0, 1].set_title("Subplot 2")
ax[0, 1].set_xlabel('SP2 x-axis')
ax[0, 1].set_ylabel('SP2 y-axis')

# Third subplot
ax[1, 0].set_title("Subplot 3")
ax[1, 0].set_xlabel('SP3 x-axis')
ax[1, 0].set_ylabel('SP3 y-axis')

# Fourth subplot

ax[1, 1].set_title("Subplot 4")
ax[1, 1].set_xlabel('SP4 x-axis')
ax[1, 1].set_ylabel('SP4 y-axis')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

### GridSpec

You can create flexible layouts by having subplots span multiple rows and/or columns using GridSpec. GridSpec allows you to define a grid for your subplots and specify which rows and columns each subplot should occupy.

```python
from matplotlib.gridspec import GridSpec

# Create a figure

fig = plt.figure(figsize=(10, 8))

fig.suptitle('Figure Title')

# Define a 3x3 grid
gs = GridSpec(3, 3, figure = fig)

# Subplot 1: Spans 2 columns (first row, all columns)
ax1 = fig.add_subplot(gs[0, :2])  # Span all columns of the first row
ax1.set_title('Subplot 1 x')

# Subplot 2: Spans 2 rows (first and second rows, third column)
ax2 = fig.add_subplot(gs[:2, 2])  # Span both rows in the third column
ax2.set_title('Subplot 2')

# Subplot 3: Standard 1x1 (second row, first column)
ax3 = fig.add_subplot(gs[1, :2])
ax3.set_title('Subplot 3')

# Subplot 4: Spans entire bottom row (third row, all columns)
ax4 = fig.add_subplot(gs[2, :])  # Span all columns in the third row
ax4.set_title('Subplot 4')

# Adjust layout
plt.tight_layout()
plt.show()
```

### Exercise (Data -> Data Extraction -> Projection)

```python
# Data

nba_data = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/NBA2324-NoTOT.csv')
nba_data.to_csv('nba.csv')
```

```python
# 01. Create two histograms comparing the FT% of NYK and MIA (use subplots)

nyk_data = nba_data[nba_data['Tm'] == 'NYK']['FT%']
mia_data = nba_data[nba_data['Tm'] == 'MIA']['FT%']

# Create a 1x2 grid of subplots

plt.figure(figsize=(12, 5))
plt.suptitle('Comparative FT%')

# First subplot
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, 1st subplot
plt.title('NYK')
plt.hist(nyk_data, bins = 5, edgecolor = 'black')
plt.xlabel('FT%')
plt.ylabel('Frequency')

# 2nd subplot
plt.subplot(1, 2, 2)  # 1 rows, 2 columns, 2nd subplot
plt.title('MIA')
plt.hist(mia_data, bins = 5, color = 'red', edgecolor = 'black')
plt.xlabel('FT%')
plt.ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# INSIGHTS

# NYK (Blue Histogram):
# The histogram shows a wider spread in FT% among NYK players, suggesting variability in free throw performance.
# There are notable frequencies across various FT% ranges, which could indicate inconsistencies in free throw skills among the team members.

# MIA (Red Histogram):
# MIA playersâ€™ FT% distribution appears slightly more concentrated, with many players falling within a specific FT% range,
# possibly suggesting a more consistent free throw performance across the team.
```

```python
# 02. Compare the top 3 scorers for teams GSW and DAL using bar graphs

# Filter data for GSW (Golden State Warriors) and DAL (Dallas Mavericks)

gsw_data = nba_data[nba_data['Tm'] == 'GSW']
dal_data = nba_data[nba_data['Tm'] == 'DAL']

gsw_data = gsw_data[['Player', 'PTS']].sort_values('PTS', ascending = False).head(3)
dal_data = dal_data[['Player', 'PTS']].sort_values('PTS', ascending = False).head(3)

plt.figure(figsize=(12, 5))
plt.suptitle('Top 3 Scorers')

# First subplot
plt.subplot(1, 2, 1)  # 1 rows, 2 columns, 1st subplot
plt.title('Golden State Warriors')
plt.bar(gsw_data['Player'], gsw_data['PTS'], color = 'gold')
plt.xlabel('Players')
plt.ylabel('PTS')
plt.grid(True, alpha = 0.15)

# 2nd subplot
plt.subplot(1, 2, 2)  # 1 rows, 2 columns, 2nd subplot
plt.title('Dallas Mavericks')
plt.bar(dal_data['Player'], dal_data['PTS'], color = 'blue')
plt.xlabel('Players')
plt.ylabel('PTS')
plt.grid(True, alpha = 0.15)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# INSIGHTS

# Reliance on Primary Scorers: Both GSW and DAL appear to rely heavily on a main player for scoring.

# Potential for Supporting Players: Each teamâ€™s second and third top scorers have room to potentially increase their
# scoring contributions, which might add depth to the teamâ€™s offensive strategy.

# Strategic Insights: Opponents could consider defensive strategies focused on limiting the influence of these primary scorers,
# potentially forcing the secondary scorers to step up.
```

```python
# 03. Create pie charts on PTS, AST, TRB for GSW, Top 5 per category

Team = 'GSW'
top = 5

top_pts = nba_data[nba_data['Tm'] == Team][['Player', 'PTS']].sort_values('PTS', ascending = False).head(top)
top_ast = nba_data[nba_data['Tm'] == Team][['Player', 'AST']].sort_values('AST', ascending = False).head(top)
top_trb = nba_data[nba_data['Tm'] == Team][['Player', 'TRB']].sort_values('TRB', ascending = False).head(top)

# explode = [0] * top
explode = [0.1] + [0] * (top - 1)  # Explode

plt.figure(figsize=(12, 12))
plt.suptitle(f'{Team} - Top {top} per category', y = 1)

# First subplot
plt.subplot(3, 1, 1)  # 1 rows, 1 columns, 1st subplot
plt.title('Points')
patches, texts, autotexts = plt.pie(top_pts['PTS'], labels = top_pts['Player'], autopct='%.2f%%', explode = explode)

# 2nd subplot
plt.subplot(3, 1, 2)  # 2 rows, 1 columns, 2nd subplot
plt.title('Assists')
plt.pie(top_ast['AST'], labels = top_ast['Player'], autopct='%.2f%%', explode = explode)

# 3rd subplot
plt.subplot(3, 1, 3)  # 3 rows, 1 columns, 3rd subplot
plt.title('Rebounds')
plt.pie(top_trb['TRB'], labels = top_trb['Player'], autopct='%.2f%%', explode = explode)

plt.axis('equal') # Ensure the pie chart is drawn as a circle

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# LAL Lakers

# Create sample data to display in each plot type
# Filter for a specific team or characteristic to make each plot type relevant and meaningful
team_data = nba_data[nba_data['Tm'] == 'LAL']  # Using Los Angeles Lakers (LAL) as an example

# Top 5 players by Points (for bar chart and pie chart)
top_pts = team_data.nlargest(5, 'PTS')

# Sample data for line and scatter plot
games_played = team_data['G'].sort_values(ascending = True)
points_per_game = team_data['PTS'].sort_values(ascending = True)
assists_per_game = team_data['AST'].sort_values(ascending = True)

# Plotting a 2x2 grid with different plot types
fig, ax = plt.subplots(2, 2, figsize=(15, 10), dpi=100)

fig.suptitle("Los Angeles Lakers")

# 1. Bar chart - Top 5 players by points
ax[0, 0].bar(top_pts['Player'], top_pts['PTS'], color='purple')
ax[0, 0].set_title("Top 5 Scorers - LAL")
ax[0, 0].set_xlabel("Player")
ax[0, 0].set_ylabel("Points Per Game")

# 2. Pie chart - Distribution of points among top 5 players
ax[0, 1].pie(top_pts['PTS'], labels=top_pts['Player'], autopct='%1.1f%%', startangle=140)
ax[0, 1].set_title("Points Distribution - Top 5 Scorers (LAL)")

# 3. Line chart - Points progression over games (sample)
ax[1, 0].plot(assists_per_game, points_per_game, color='blue', marker='o', linestyle='-', linewidth=1)
ax[1, 0].set_title("Points per Game (All Players) - LAL")
ax[1, 0].set_xlabel("Assists Per Game")
ax[1, 0].set_ylabel("Points Per Game")

# 4. Scatter plot - Assists vs Points (sample)
ax[1, 1].scatter(games_played, assists_per_game, color='red', alpha=0.7)
ax[1, 1].set_title("Games Played vs Assists - LAL")
ax[1, 1].set_xlabel("Games Played")
ax[1, 1].set_ylabel("Assists Per Game")

plt.tight_layout()
plt.show()
```

```python
# Filter for a specific team, e.g., "LAL" (Los Angeles Lakers)
team_data = nba_data[nba_data['Tm'] == 'LAL']

# Create subplots
fig, ax = plt.subplots(2, 2, figsize=(15, 12), dpi=100)

# 1. Boxplot of Points (PTS) for the specific team
ax[0, 0].boxplot(team_data['PTS'].dropna())
ax[0, 0].set_title('Boxplot of Points (PTS) - LAL')
ax[0, 0].set_ylabel('Points (PTS)')

# 2. Violin plot of Assists (AST) for the specific team
violin_parts = ax[0, 1].violinplot(team_data['AST'].dropna(), showmeans=True, showmedians=True)
violin_parts['cmeans'].set_color('black')  # Set mean color to blue
violin_parts['cmedians'].set_color('red')  # Set median color to red

ax[0, 1].set_title('Violin Plot of Assists (AST) - LAL')
ax[0, 1].set_ylabel('Assists (AST)')


# 3. Hexbin plot of Assists (AST) vs Points (PTS) for the specific team
hb = ax[1, 0].hexbin(team_data['AST'].dropna(), team_data['PTS'].dropna(), gridsize=20, cmap='Blues')
ax[1, 0].set_title('Hexbin Plot of Assists (AST) vs Points (PTS) - LAL')
ax[1, 0].set_xlabel('Assists (AST)')
ax[1, 0].set_ylabel('Points (PTS)')
fig.colorbar(hb, ax=ax[1, 0], label='Frequency')

# 4. Stem plot of Points (PTS) vs Rebounds (TRB) for the specific team
# Using PTS as x-values and TRB as y-values for stem plot
x = team_data['PTS'].dropna()
y = team_data['TRB'].dropna()
ax[1, 1].stem(x, y, linefmt='--', markerfmt='o', basefmt='r')
ax[1, 1].set_title('Stem Plot of Points (PTS) vs Rebounds (TRB) - LAL')
ax[1, 1].set_xlabel('Points (PTS)')
ax[1, 1].set_ylabel('Rebounds (TRB)')

plt.tight_layout()
plt.show()
```

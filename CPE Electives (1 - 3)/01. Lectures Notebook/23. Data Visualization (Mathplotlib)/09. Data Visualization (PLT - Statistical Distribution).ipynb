### Error Bar

Used to plot data with error bars, which visually represent the variability or uncertainty of each data point. Error bars can indicate standard deviations, confidence intervals, or any other measures of variability around the data points, helping viewers understand the reliability of the data.

```python
# Generate sample data
x = np.linspace(0, 10, 10)               # X-axis values
y = np.sin(x)                            # Y-axis values (data points)
y_error = 0.1 + 0.1 * np.sqrt(x)         # Variable error margin

plt.figure(figsize = (12, 5))

# Create error bar plot
plt.errorbar(x, y, yerr = y_error, fmt='o', color='blue', ecolor='red', capsize=5, label='Data with Error Bars')

# Customizations
plt.title('Error Bar Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# Show plot
plt.show()

# x and y: The x and y coordinates of the data points.
# yerr: Defines the vertical error bars for each point. This can be an array of error values
#       (like standard deviations) or a single scalar for uniform errors.
# fmt: Specifies the marker format for the data points ('o' for circular markers).
# color: Sets the color of the data points.
# ecolor: Sets the color of the error bars.
# capsize: Adds caps at the end of the error bars, making them easier to see.
# label: Adds a label for the legend.


# Purpose of Error Bars

# Visualizing Data Uncertainty: Error bars help indicate the uncertainty, variance, or possible range of each data point.
# Assessing Reliability: They allow viewers to assess the reliability and consistency of measurements.
# Comparative Analysis: Error bars make it easier to compare different datasets and understand overlapping values,
#                       especially in scientific and statistical plots.
```

### Violin Plot

Used to visualize the distribution of a dataset across different categories. It combines aspects of a box plot and a density plot, showing both the range and the probability density of the data. Violin plots are especially useful when comparing the distribution of several datasets or categories.

```python
# Generate sample data
np.random.seed(10)  # Set seed for reproducibility
data1 = np.random.normal(0, 1, 100)  # Dataset 1 with mean 0, std dev 1
data2 = np.random.normal(1, 1.5, 100)  # Dataset 2 with mean 1, std dev 1.5
data3 = np.random.normal(2, 0.5, 100)  # Dataset 3 with mean 2, std dev 0.5

# Organize data for plotting
data = [data1, data2, data3]

plt.figure(figsize = (12,5))

# Create a violin plot
violin_parts = plt.violinplot(data, showmeans=True, showmedians=True, vert = False) # vert = False

violin_parts['cmeans'].set_color('blue')  # Set mean color to blue
violin_parts['cmedians'].set_color('red')

# Customizations
plt.title('Violin Plot Example')
plt.ylabel('Category')
plt.xlabel('Values')
plt.yticks([1, 2, 3], ['Category 1', 'Category 2', 'Category 3'])  # Label the categories
plt.grid(True)

# Show plot
plt.show()

# data: A list of arrays, where each array represents a category (e.g., data distributions for different groups).
# showmeans: If set to True, shows the mean value of each dataset as a point on the plot.
# showmedians: If set to True, shows the median value of each dataset as a line across the violin.
# xticks: Used to label the categories on the x-axis

# Purpose of Violin Plots

# Distribution Visualization: Violin plots provide a smooth representation of the data distribution, making it easier to see
#                             where values are concentrated.

# Comparing Multiple Datasets: They are particularly useful for comparing the distributions of multiple datasets or categories.

# Outliers and Variability: Violin plots allow for insights into data spread, outliers, and differences in distribution shape,
#                           which may not be evident in box plots.
```

```python
# In comparison to Box Plot

# Generate sample data
np.random.seed(10)
data1 = np.random.normal(0, 1, 100)  # Dataset 1 with mean 0, std dev 1
data2 = np.random.normal(1, 1.5, 100)  # Dataset 2 with mean 1, std dev 1.5
data3 = np.random.normal(2, 0.5, 100)  # Dataset 3 with mean 2, std dev 0.5

# Organize data for plotting
data = [data1, data2, data3]

# Create a horizontal box plot
plt.figure(figsize=(12, 5))
plt.boxplot(data, vert=False, showmeans=True, meanline=True)

# Customizations
plt.title('Horizontal Box Plot Example')
plt.xlabel('Values')
plt.ylabel('Category')
plt.yticks([1, 2, 3], ['Category 1', 'Category 2', 'Category 3'])  # Label the categories
plt.grid(True)

# Show plot
plt.show()
```

### Event Plot

Used to create event plots, also known as "spike plots." These plots are useful for visualizing the timing of events, such as occurrences in time series data, or for displaying sequences of events across multiple categories or trials. They are often used in neuroscience, experiment timing analysis, and other fields where understanding event distribution is important.

```python
# Generate sample event data
np.random.seed(0)
data1 = np.random.rand(10) * 10   # Event times for category 1
data2 = np.random.rand(15) * 10   # Event times for category 2
data3 = np.random.rand(20) * 10   # Event times for category 3

plt.figure(figsize = (12,5))

# Organize data for plotting
data = [data1, data2, data3]

# Create an event plot
plt.eventplot(data, orientation='horizontal', lineoffsets=[1, 2, 3], linelengths=0.8, colors=['blue', 'green', 'purple'])

# Customizations
plt.title('Event Plot Example')
plt.xlabel('Time')
plt.ylabel('Test Trials')
plt.yticks([1, 2, 3], ['Trial 1', 'Trial 2', 'Trial 3'])  # Label the categories
plt.grid(True)

# Show plot
plt.show()

# data: A list of arrays, where each array contains the event times for a category. Each array is plotted as a series of lines or spikes.
# orientation: 'horizontal' or 'vertical', specifying the orientation of the event lines.
# lineoffsets: Specifies the y-offsets for each category of events, making it possible to stack different categories.
# linelengths: Controls the length of each event line. Here, 0.8 makes them slightly shorter than the default.
# colors: Sets the color for each category in the plot.
```

### Hist2D

Used to create two-dimensional histograms. This type of plot is useful for visualizing the distribution of values across two variables, similar to a heatmap. It divides the space into bins and counts the number of occurrences in each bin, making it ideal for identifying density patterns and correlations between two variables.

```python
# Generate sample data
np.random.seed(0)
x = np.random.randn(1000)  # Normally distributed data for x-axis
y = np.random.randn(1000)  # Normally distributed data for y-axis

plt.figure(figsize = (12, 5))

# Create a 2D histogram
plt.hist2d(x, y, bins=30, cmap='plasma')

# Customizations
plt.title('2D Histogram (hist2d) Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Counts')  # Add color bar to indicate counts

# Show plot
plt.show()

# x and y: Arrays representing the data points along the x and y axes.
# bins: The number of bins in each dimension (30x30 grid here), controlling the plot's resolution.
# cmap: Sets the color map, which helps visualize density (e.g., 'Blues' for shades of blue, where darker shades represent higher densities).

# Purpose of 2D Histograms

# Density Visualization: Shows the distribution and density of data points in two dimensions.
# Finding Correlations: Useful for detecting relationships between two variables, as denser regions indicate stronger correlations.
# Analyzing Data Clusters: Helps identify clusters, gaps, and other patterns that may not be obvious in scatter plots.
```

### Hexbin

Used to create hexagonal binning plots, which are similar to 2D histograms but use hexagon-shaped bins instead of rectangular ones. Hexagonal binning is particularly effective for visualizing the density of data points when there’s a large dataset and can reveal patterns more effectively than a standard scatter or 2D histogram plot.

```python
# Generate sample data
np.random.seed(0)
x = np.random.randn(10000)  # Normally distributed data for x-axis
y = np.random.randn(10000)  # Normally distributed data for y-axis

plt.figure(figsize = (12, 5))

# Create a hexbin plot
plt.hexbin(x, y, gridsize=30, cmap='YlGnBu', mincnt=1)

# Customizations
plt.title('Hexbin Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Counts')  # Add color bar to indicate counts

# Show plot
plt.show()

# x and y: Arrays representing the data points along the x and y axes.
# gridsize: Controls the number of hexagons in the plot (the resolution of the hexbin). Larger gridsize values create smaller hexagons.
# cmap: Specifies the color map (e.g., 'YlGnBu' for yellow-green-blue), where darker shades represent higher densities.
# mincnt: Sets a minimum count to color bins, useful to avoid very light colors in bins with few counts.
```

### ECDF (Empirical Cumulative Distribution Function)

Used to show the cumulative distribution of a dataset. Unlike a histogram or kernel density estimation (KDE), which shows the probability density, an ECDF shows the proportion of data points that are below each x-value. ECDFs are useful for comparing distributions between datasets and seeing percentile information directly on the plot.

```python
# Generate sample data
np.random.seed(0)
data = np.random.normal(0, 1, 100)  # Normally distributed data with mean 0 and std dev 1

plt.figure(figsize = (12, 5))

# Sort data and calculate ECDF values
x = np.sort(data)                             # Sorted data for x-axis
y = np.arange(1, len(data) + 1) / len(data)   # ECDF values for y-axis (proportion of data points)

# Plot ECDF
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='.', linestyle='none', color='blue')
plt.title('ECDF Plot Example')
plt.xlabel('Data Values')
plt.ylabel('ECDF')
plt.grid(True)

# Show plot
plt.show()

# x: The sorted data values.
# y: Calculated ECDF values as np.arange(1, len(data) + 1) / len(data).
#    This gives the cumulative proportion of data points up to each x value.
```

```python
# Generate sample data for two datasets
np.random.seed(0)
data1 = np.random.normal(0, 1, 100)    # Dataset 1: mean 0, std dev 1
data2 = np.random.normal(1, 1.5, 100)  # Dataset 2: mean 1, std dev 1.5

# Calculate ECDF values for Dataset 1
x1 = np.sort(data1)
y1 = np.arange(1, len(data1) + 1) / len(data1)

# Calculate ECDF values for Dataset 2
x2 = np.sort(data2)
y2 = np.arange(1, len(data2) + 1) / len(data2)

# Plot ECDFs
plt.figure(figsize=(12, 5))
plt.plot(x1, y1, marker='.', linestyle='none', color='blue', label='Dataset 1')
plt.plot(x2, y2, marker='.', linestyle='none', color='red', label='Dataset 2')

# Customizations
plt.title('Comparison of ECDFs for Two Datasets')
plt.xlabel('Data Values')
plt.ylabel('ECDF')
plt.legend()
plt.grid(True)

# Show plot
plt.show()

# x1, y1: ECDF values for Dataset 1. x1 is sorted data, and y1 is calculated as the cumulative proportion of data points.
# x2, y2: ECDF values for Dataset 2, calculated similarly.
# Overlaying ECDFs: By plotting both ECDFs on the same plot with different colors ('blue' and 'red'),
#                   we can visually compare the distributions.

# Purpose and Insights

# Comparing Distributions: This plot allows you to quickly see differences in the central tendency, spread, and shape of the two datasets.
# Percentile Comparison: You can easily identify percentiles (e.g., median, quartiles) and see if one
#                        dataset has higher or lower values at certain percentiles.
# Visualizing Overlaps and Gaps: Areas where the ECDFs are close or diverge can indicate similarities or differences in the datasets.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(0)
data = np.random.normal(0, 1, 100)  # Normally distributed data

# Sort data and calculate ECDF values
x = np.sort(data)                            # Sorted data values for x-axis
y = np.arange(1, len(data) + 1) / len(data)  # ECDF values for y-axis

# Calculate percentiles
percentiles = [25, 50, 75]
percentile_values = np.percentile(data, percentiles)  # Values at the 25th, 50th, and 75th percentiles

# Plot ECDF
plt.figure(figsize=(12, 5))
plt.plot(x, y, marker='.', linestyle='none', color='blue', label='ECDF')
plt.title('ECDF Plot with Percentile Lines')
plt.xlabel('Data Values')
plt.ylabel('ECDF')
plt.grid(True)

# Add percentile lines
for i, percentile in enumerate(percentiles):

    # Horizontal line for percentile
    plt.axhline(y=percentile / 100, color='gray', linestyle='--', linewidth=1)
    # Vertical line at the percentile value
    plt.axvline(x=percentile_values[i], color='red', linestyle='--', linewidth=1)
    # Annotation
    plt.text(percentile_values[i] - 0.35, percentile / 100 + 0.005, f'{percentile}%', color='black', ha='left', va='bottom')

plt.legend()
plt.show()

# Explanation of Percentile Lines

# percentiles: A list of percentiles we want to mark (e.g., 25th, 50th, and 75th).
# np.percentile(data, percentiles): Calculates the data values at each percentile.
# axhline and axvline: Adds horizontal lines at the y-axis values corresponding to the percentiles and vertical lines at the x-axis values of each percentile.
# Annotations: Labels each line with its percentile value to improve clarity.

# Purpose of Adding Percentile Lines

# Enhanced Interpretability: Percentile lines help quickly locate specific percentile values on the plot.
# Data Insights: The 25th, 50th, and 75th percentiles give insights into data spread and central tendencies.
```

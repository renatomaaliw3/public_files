### Stem Plot

A stem plot is useful for displaying data points on discrete intervals along with their magnitudes,
typically resembling a series of "stems" coming out of the baseline.

```python
# Generate sample data

x = np.linspace(0, 2 * np.pi, 50)  # 50 points from 0 to 2Ï€
y = np.sin(x)  # Sinusoidal data for y-axis

plt.figure(figsize = (12,5))

# Create a stem plot
plt.stem(x, y, linefmt='g-', markerfmt='o', basefmt='red')

# Customizations
plt.title('Sample Stem Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Show plot
plt.show()

# x and y: These are the data points. x is the position, and y is the magnitude of the stems.
# linefmt: Specifies the style of the stem lines (-- is for dashed lines).
# markerfmt: Specifies the style of the markers at the tip of each stem (o is for circular markers).
# basefmt: Specifies the style of the baseline (e.g., r for a red line)
```

### Fill Between

The fill_between function in Matplotlib is used to fill the
area between two horizontal curves or between a curve and a baseline (usually the x-axis). This can be useful for visually emphasizing the area under a curve, showing ranges of values, or highlighting specific regions on a plot.

```python

# Generate sample data
x = np.linspace(0, 10, 100)       # X-axis values from 0 to 10
y1 = np.sin(x)                    # First curve: sine wave
y2 = np.cos(x)                    # Second curve: cosine wave

plt.figure(figsize = (12,5))

# Create plot
plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='red')

# Fill area between y1 and y2
plt.fill_between(x, y1, y2, color = 'gray', alpha = 0.3)

# Customizations
plt.title('Area Between Sine and Cosine')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# Show plot
plt.show()

# The fill_between function fills the area between y1 and y2 in a gray color, which makes
# it easy to see the regions where y1 is above or below y2.
```

```python
# Generate sample data
x = np.linspace(0, 10, 100)  # X-axis values from 0 to 10
y = np.sin(x)                # Sine wave data

plt.figure(figsize = (12,5))

# Plot the curve
plt.plot(x, y, label='sin(x)', color='blue')

# Fill area between the curve and the x-axis (y=0)
plt.fill_between(x, y, 0, where = (y > 0), color = 'green', alpha = 0.3, label='Positive Area')
plt.fill_between(x, y, 0, where = (y < 0), color = 'red', alpha = 0.3, label='Negative Area')

# Customizations
plt.title('Shaded Area Under Sine Curve')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

### Stack Plot

This type of plot is ideal for visualizing how different components contribute to a total over time, making it especially useful for showing cumulative data distributions or resource usage.

```python
# Generate sample data
x = np.linspace(0, 10, 100)             # X-axis values (e.g., time)
y1 = np.sin(x) + 2                      # First data series
y2 = np.cos(x) + 2                      # Second data series
y3 = np.sin(x / 2) + 1.5                # Third data series

plt.figure(figsize = (12,5))

# Stack plot
plt.stackplot(x, y1, y2, y3, labels=['Data 1', 'Data 2', 'Data 3'], colors=['skyblue', 'orange', 'lightgreen'])

# Customizations
plt.title('Stacked Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best')
plt.grid(True)

# Show plot
plt.show()
```

### Stairs

Commonly used for data that changes at specific intervals,
such as time series data with discrete steps or histogram-like data representations.

```python
# Generate sample data
x = np.linspace(0, 10, 11)  # X-axis values (e.g., time intervals or categories)
y = np.random.randint(1, 10, size=10)  # Y-axis values for each interval

plt.figure(figsize = (12,5))

# Create a stairs plot
plt.stairs(y, x, label='Step Data', color='purple', linewidth=2)

# Customizations
plt.title('Stairs Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

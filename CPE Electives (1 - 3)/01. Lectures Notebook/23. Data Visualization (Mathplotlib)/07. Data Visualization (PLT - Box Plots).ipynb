#**Box Plot**

A box plot (or box-and-whisker plot) is a graphical representation of data distribution based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It is widely used in descriptive statistics to provide a quick visual summary of the data, highlighting its central tendency, spread, and potential outliers.

##**Purpose of Box Plot**:

  -- _**Summarize Data Distribution**_: Box plots quickly reveal the center, spread, and symmetry (or skewness) of the data.

  -- _**Identify Outliers**_: They are especially useful for spotting outliers, which can highlight interesting data points, measurement errors, or irregularities in data.

  -- _**Compare Multiple Groups:**_: Box plots are ideal for comparing the distributions of several groups side-by-side, showing differences in spread, central tendency, and outlier presence.

  -- _**Detect Outliers**_: Outliers or unusual values are easily identified in a histogram, as they often fall into sparse or empty bins on either end of the chart.

  -- _**Data Insights**_: Box plots help in assessing the degree of data spread, concentration, and skewness, which can guide data transformations, normalization, or additional exploratory analyses.

```python
# Histogram with KDE (Seaborn)

plt.figure(figsize = (12, 5))

# Histogram
sns.histplot(df['normal_dist'], bins = 30, color = 'green', edgecolor = 'black', label = 'Distribution', kde = True)
plt.title("Basic Histogram and KDE")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.legend()
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# A2. Basic Box Plot (Customization)

plt.figure(figsize = (12, 5))

# Boxplot
plt.boxplot(df['normal_dist'], patch_artist = True,
                               boxprops = dict(facecolor = 'red', color = 'blue', linewidth = 3),
                               medianprops = dict(color = 'green', linewidth = 3),
                               whiskerprops = dict(color = 'orange'),
                               capprops = dict(linewidth = 2, color = 'red'),
                               flierprops = dict(marker = 'o', color = 'red', markersize = 12))


plt.title("Box Plot (Normal Distribution)")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# A3. Basic Box Plot (Multiples)

plt.figure(figsize = (12, 5))

# Boxplot
plt.boxplot([df['normal_dist'], df['uniform_dist'], df['exponential_dist']], labels = ['Normal', 'Uniform', 'Exponential'], vert = False,
                                patch_artist = True, boxprops = dict(facecolor = 'lightblue', color = 'black', linewidth = 2),
                                medianprops = dict(color = 'green', linewidth = 3))

plt.title("Box Plot (Normal Distribution)")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# How to identify the outliers

# Basic Box Plot (Vertical, Horizontal)
plt.figure(figsize = (12, 5))

# Boxplot
plt.boxplot(df['normal_dist']) # vert = False
plt.title("Box Plot (Normal Distribution)")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

```python
# Identifying outliers in the `normal_dist` column based on the IQR method
# Outliers are values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR

# Calculate Q1, Q3, and IQR for normal_dist
df_normal = df['normal_dist']

Q1 = df_normal.quantile(0.25)
Q3 = df_normal.quantile(0.75)
IQR = Q3 - Q1

# Define the outlier thresholds
lower_bound = Q1 - 1.49 * IQR
upper_bound = Q3 + 1.49 * IQR

# Identify outliers

outliers = df_normal[(df_normal < lower_bound) | (df_normal > upper_bound)]
outliers
```

```python
# Graph again without the outliers

data_cleaned = df_normal[(df_normal >= lower_bound) & (df_normal <= upper_bound)]
data_cleaned.rename('Normal')

# Basic Box Plot (Vertical, Horizontal)
plt.figure(figsize = (12, 5))

# Boxplot
plt.boxplot(data_cleaned) # vert = False
plt.title("Box Plot (Normal Distribution no Outliers)")
plt.ylabel("Y - Values")
plt.xlabel("Distribution")
plt.grid(True, alpha = 0.25, color = 'red')
plt.show()
```

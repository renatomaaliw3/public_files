# Seaborn

```
import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
%matplotlib inline
```

## Scatter Plot

```
# Data

tips = sns.load_dataset('tips')
tips
```

```
# Scatter Plot is also Relationship Plot

sns.relplot(x = 'total_bill', y = 'tip', data = tips, color = 'r')
plt.title('Tips')
```

### Hue, Style and Size

```
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time', size = 'size')
plt.title('Tips')
```

```
# Sizing Plots in RelPlot

sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size', sizes = (15,200))
plt.title('Tips')
```

# Line Plot

```
# Generate Data

np.random.seed(10)

df = pd.DataFrame(dict(time = np.arange(500), value = np.random.randn(500).cumsum()))
df
```

```
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df, linestyle = '--', color = 'r')
```

```
# Data

fmri = sns.load_dataset('fmri')
fmri.head()
```

```
# Line Plot with Sine (Probability)

sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, label = 'Sine')
```

```
# Line Plot with Sine (Probability), Height, Aspect (Length)

sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, label = 'Signal', height = 4, aspect = 2)
plt.legend(loc = 'center right')
```

```
# Multiple Lines with Markers

sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', hue = 'event', style = 'region', data = fmri, height = 5, aspect = 3, markers = True)
```

### Suplots

```
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time', row = 'sex', data = tips)
```

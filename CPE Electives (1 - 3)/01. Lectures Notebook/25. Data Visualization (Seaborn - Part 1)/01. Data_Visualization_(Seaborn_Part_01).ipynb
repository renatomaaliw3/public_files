# Statistical Data Visualization (Seaborn) - Relationship

      - Seaborn is a Python data visualization library based on matplotlib.
      - It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
# Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pdb
```

## Relationship Plots | .relplot()

```python
# Dataset

tips = sns.load_dataset("tips")
tips.head()
```

```python
# Relplot()

sns.relplot(data = tips, x = "total_bill", y = "tip")
plt.grid(True)
```

```python
# Relplot(), "hue semantic"

sns.relplot(data = tips, x = "total_bill", y = "tip", hue = 'smoker')
plt.grid(True)
```

```python
# Relplot(), "hue semantic" and "style" semantic

sns.relplot(data = tips, x = "total_bill", y = "tip", hue = 'smoker', style = 'time')
plt.grid(True)
```

```python
# Relplot(), "hue semantic" and "style" semantic, with pallette and sizes

sns.relplot(data = tips, x = "total_bill", y = "tip", hue = 'smoker',
            style = 'time', palette = 'Greens', size = 'size', sizes = (50, 200))
plt.grid(True)
```

## Relationship Plot | kind = 'line'

```python
# Data

dowjones = sns.load_dataset('dowjones')
dowjones.head()
```

```python
# Relplot, kind = 'line'

sns.relplot(data = dowjones, x = "Date", y = "Price", kind = "line")
```

```python
# Data

fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
# Line with uncertainty based on data

sns.relplot(data = fmri, x = "timepoint",
            y = "signal", kind = "line", errorbar = 'sd', estimator = None)

# errobar = None, errobar = 'sd' (standard deviation), estimator = None
```

```python
# Line with semantics, hue

sns.relplot(data = fmri, x = "timepoint", y = "signal", kind = "line", hue = 'event')
```

```python
# Line with semantics, style

sns.relplot(data = fmri, x = "timepoint", y = "signal",
            kind = "line", hue = 'region', style = 'event', legend = True)
```

```python
# Line with markers

sns.relplot(data = fmri, x = "timepoint", y = "signal", kind = "line",
            hue = 'region', style = 'event', markers = True)
```

```python
# Showing multiple relationships with facets

g = sns.relplot(data = tips, x = "total_bill", y = "tip", hue = "smoker", col = "day")

# Add gridlines to all subplots

for ax in g.axes.flat:

    ax.grid(True, linestyle = '--', alpha = 0.75)  # Enable grid for each subplot

# Display the plot
plt.show()
```

```python
# Showing multiple relationships with facets

marker_mapping = {'stim': 'o', 'cue': 'X'}
colors = {'blue', 'red'}

g = sns.relplot(
    data = fmri.query("region == 'frontal'"), kind="line",
    x = "timepoint", y = "signal", hue = "event", style = "event",
    col = "subject", col_wrap = 3,
    height = 3, aspect = 1, linewidth = 2.5, markers = marker_mapping, palette = colors
)

# Add gridlines to all subplots

for ax in g.axes.flat:

    ax.grid(True, linestyle = '--', alpha = 0.75)  # Enable grid for each subplot


# Add a title
g.fig.suptitle('Signal Analysis', fontsize = 18, y = 1.05)

# Adjust layout to prevent overlap
# g.fig.tight_layout()

plt.show()
```

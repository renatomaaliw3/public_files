# Hierarchical Clustering

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Data

```python
df = pd.read_csv('Mall_Customers.csv')
df.head(3)
```

```python
# reduced_df

reduced_df = df.drop(['Gender','Age', 'CustomerID'], axis = 1)
reduced_df.head(3)
```

## Dendograms

```python
# Import

import scipy.cluster.hierarchy as sch

plt.rcParams['figure.figsize'] = (8, 6)

dendrogram = sch.dendrogram(sch.linkage(reduced_df, method = 'ward'))
plt.title('Dendogram', fontsize = 20)
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

## Hierarchy Clustering


```python
# Import

from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(reduced_df)
```

```python
# Plot

sns.scatterplot(data = reduced_df, x = reduced_df.iloc[:,0], y = reduced_df.iloc[:,1], hue = y_hc, palette = ['red','blue','green','yellow','brown'], s = 80).set(title = 'Customer Clusters')
plt.grid()
```

# SQL in Python

```python
# Imports

import sqlite3
import pandas as pd
```

```python
# Load data

data = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/master/Data%20Sets/summer.csv')
data.head(5)
```

```python
# Connect to an in memory SQLite Database

conn = sqlite3.connect(':memory')
```

```python
# Write the DataFrame to the Database

data.to_sql('Summer', conn, index = False)
```

```python
# Execute a SQL Query

query = "SELECT COUNT(DISTINCT Athlete) AS Names FROM Summer WHERE Event = '100M Freestyle'"
result_sql = pd.read_sql_query(query, conn)
```

```python
# Display the Query Results

result_sql
```

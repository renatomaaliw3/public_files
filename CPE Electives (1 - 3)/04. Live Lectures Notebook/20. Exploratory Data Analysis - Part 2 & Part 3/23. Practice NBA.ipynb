# NBA Data

```python
# Imports

import numpy as np
import pandas as pd
```

```python
# Load Dataset

nba = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/master/Data%20Sets/nba.csv')
nba
```

```python
# Check for missing data

nba.isnull().sum()
```

```python
# Remove data with all column values of 'NaN'

nba = nba.dropna(how = 'all') # all column values with missing data
nba
```

```python
# Replace Salary with NaN with 0

nba['Salary'].fillna(0, inplace = True)
nba
```

```python
# Replace College with NaN with Uknown

nba['College'].fillna('Unknown', inplace = True)
nba
```

```python
# Show actual rows with missing data

nba[nba.isnull().any(axis = 1)]
```

```python
# Remove all data with NaN

nba.dropna(inplace = True)
nba
```

```python
# astype

nba['Salary'] = nba['Salary'].astype('int')
nba
```

```python
# Reset Index

nba.reset_index(drop = True, inplace = True)
nba
```

## Data Extraction Practice

```python
# 1. What is the average salary of NBA players

nba['Salary'].mean()
```

```python
# 2. Who are the top 3 highest paid player on the Golden State Warriors roster?

nba[nba['Team'] == 'Golden State Warriors'].sort_values(by = 'Salary', ascending = False).head(3)
```

```python
# 3. Who are the highest paid player for each team in the NBA, sort by Salary Descending?

nba.loc[nba.groupby('Team')['Salary'].idxmax()][['Team', 'Name', 'Salary']].sort_values('Salary', ascending = False)
```

```python
# 4. How many players are part of each Team?

nba.groupby('Team').count()['Name']
```

```python
# 5. Who are the players that have salaries between 20M to 40M in the Los Angeles Lakers and Los Angeles Clippers, in Salary descending order?

nba.query("(Salary >= 20000000 and Salary <= 60000000) and Team in ['Los Angeles Lakers', 'Los Angeles Clippers']").sort_values('Salary', ascending = False)
```

```python
# 6. The General Manager is looking for a G (Guard) with weight of 200 lbs. above with current salary of 10M and below

is_guard = nba['Position'] == 'G'
req_weight = nba['Weight'] == 200
req_salary = nba['Salary'] <= 10000000

nba[is_guard & req_weight & req_salary]
```

```python
# 7. Who are the playes with 'State' instances in their College and current Team of Utah Jazz?

with_state = nba['College'].str.contains('State', case = False)
current_team = nba['Team'] == 'Utah Jazz'

nba[with_state & current_team]
```

```python
# 8. Who are the players who entered the league from High School and has a salary of 30M and above

from_highschool = nba['College'].str.contains('HS', case = False)
req_salary = nba['Salary'] >= 20000000

nba[from_highschool & req_salary]
```

```python
# 9. The team is looking for a pure Center (C) with weight of 250 lbs - 310 lbs and a Salary of 10M to 20M, Name in Ascending, Salary in Descending

req_pos = nba['Position'] == 'C'
req_weight = (nba['Weight'] >= 250) & (nba['Weight'] <= 310)
req_salary = (nba['Salary'] >= 10000000) & (nba['Salary'] >= 20000000)

nba[req_pos & req_weight & req_salary].sort_values(by = ['Name', 'Salary'], ascending = [True, False])
```

```python
# 10. Give me players that plays two-positions from the Dallas Mavericks and San Antonio Spurs, sort by Salar Descending and Weight Descending

req_pos = nba.query("~Position.isin(['C', 'F', 'G']) & (Team == 'Dallas Mavericks' or Team == 'San Antonio Spurs')")
req_pos.sort_values(by = ['Salary', 'Weight'], ascending = [False, False])
```
